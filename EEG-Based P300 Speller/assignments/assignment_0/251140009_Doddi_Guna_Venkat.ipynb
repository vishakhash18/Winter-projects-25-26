5. Write a function definition named is_odd that takes in a number and returns True or False if that number is odd.

```python
def is_odd(num):
    if num % 2 != 0:
        return True
    else:
        return False

print("Is 7 odd?",is_odd(7))
print("Is 10 odd?",is_odd(10))
```

6. Write a function definition named mode that takes in sequence of numbers and returns the most commonly occuring value

```python
def mode(num_list):
    frequency = {}
    for num in num_list:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1
    max_count = max(frequency.values())
    most_common_value = [num for num, count in frequency.items() if count == max_count]
    return most_common_value[0] # Return the first mode found

sample_list = [4, 1, 2, 2, 3, 4, 4, 5, 1, 2, 4]
print("The mode of the sample list is:", mode(sample_list))
```

Exercise : Read face cream and facewash product sales data and show it using the bar chart \\
The bar chart should display the number of units sold per month for each product. \\
Add a separate bar for each product in the same chart.

```python
import numpy as np
import matplotlib.pyplot as plt

months = df['month_number']
facecream = df['facecream']
facewash = df['facewash']

x = np.arange(1,len(months)+1)    # positions for bars
width = 0.4                       # width of each bar

plt.bar(x - width/2, facecream, width=width, label="FACECREAM")
plt.bar(x + width/2, facewash,  width=width, label="FACEWASH")
plt.xlabel('month number')
plt.ylabel('rate')
plt.legend()
plt.show()
```
