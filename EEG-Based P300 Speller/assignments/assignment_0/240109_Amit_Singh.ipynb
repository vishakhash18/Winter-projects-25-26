
6. Write a function definition named mode that takes in sequence of numbers and returns the most commonly occuring value

```python
def mode(seq):
  count={}
  for i in seq:
    if i in count:
      count[i]+=1
    else:
      count[i]=1
  max_count=0
  mode=0
  for j,count in count.items():
    if count>max_count:
      mode=j
      max_count=count
  return mode

seq=[1,2,2,3,3,3,4,4,4,4]
print(mode(seq))
```


Exercise 1: Read Total profit of all months and show it using a line plot \\
Total profit data provided for each month. Generated line plot must include the following properties: –

X label name = Month Number \\
Y label name = Total profit \\




```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('company_sales_data.csv')
plt.style.use("dark_background")
plt.plot(data['month_number'],data['total_profit'])
plt.title("Line plot")
plt.xlabel("Month_Number")
plt.ylabel("Total_Profit")
plt.show()
```


Exercise 2: Read face cream and facewash product sales data and show it using the bar chart \\
The bar chart should display the number of units sold per month for each product. \\
Add a separate bar for each product in the same chart.

```python
month_number=data['month_number']
facecream_sales=data['facecream']
facewash_sales=data['facewash']
bar_widht=0.35
plt.bar(month_number-bar_widht/2,facecream_sales,bar_widht,color="tab:blue",label='facecream')
plt.bar(month_number+bar_widht/2,facewash_sales,bar_widht,color="tab:orange",label='facewash')
plt.xlabel("Month_Number")
plt.ylabel("rate")
plt.legend()
plt.show()
```
