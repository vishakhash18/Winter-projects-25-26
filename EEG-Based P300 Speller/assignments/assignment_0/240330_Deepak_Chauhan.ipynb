5. Write a function definition named is_odd that takes in a number and returns True or False if that number is odd.

```python
def is_odd(number):
    return number % 2 != 0

# Example usage:
print(is_odd(5))  # True
print(is_odd(8))  # False
```

6. Write a function definition named mode that takes in sequence of numbers and returns the most commonly occuring value

```python
def mode(numbers):
    frequency = {}
    for num in numbers:
        frequency[num] = frequency.get(num, 0) + 1
    max_count = max(frequency.values())
    for num, count in frequency.items():
        if count == max_count:
            return num

# Example usage:
nums = [1, 2, 2, 3, 3, 3, 4]
print(mode(nums))  # 3

```

Exercise 3: Sort following NumPy array \\
Case 1: Sort array by the second row \\
Case 2: Sort the array by the second column

```python
import numpy as np

arr = np.array([[34, 43, 73],
                [82, 22, 12],
                [53, 94, 66]])

# Case 1: Sort array by the second row
sorted_by_row = arr[:, arr[1].argsort()]
print("Sorted by second row:\n", sorted_by_row)

# Case 2: Sort the array by the second column
sorted_by_col = arr[arr[:, 1].argsort()]
print("Sorted by second column:\n", sorted_by_col)
```

Exercise 2: Replace all column values which contain ?, n.a, or NaN with suitable values and print the updated dataset.:

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Automobile_data.csv")

# Step 1: Replace '?' and 'n.a' with actual NaN
df.replace(['?', 'n.a'], np.nan, inplace=True)

# Step 2: Fill NaN values
for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:      # numerical
        df[column].fillna(df[column].mean(), inplace=True)
    else:                                              # categorical
        df[column].fillna(df[column].mode()[0], inplace=True)

# Print updated dataset
print(df)
```

Exercise 1: Read Total profit of all months and show it using a line plot \\
Total profit data provided for each month. Generated line plot must include the following properties: â€“

X label name = Month Number \\
Y label name = Total profit \\




```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

profit = df['total_profit']
months = range(1, 13)

plt.plot(months, profit, marker='o', linewidth=2)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company Profit Per Month")
plt.grid(True)
plt.show()
```

Exercise : Read face cream and facewash product sales data and show it using the bar chart \\
The bar chart should display the number of units sold per month for each product. \\
Add a separate bar for each product in the same chart.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("company_sales_data.csv")

months = np.arange(1, 13)
face_cream = df['facecream']
face_wash = df['facewash']

bar_width = 0.4

plt.bar(months - bar_width/2, face_cream, width=bar_width, label="Face Cream")
plt.bar(months + bar_width/2, face_wash, width=bar_width, label="Face Wash")

plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.title("Face Cream & Face Wash Monthly Sales")
plt.xticks(months)
plt.legend()
plt.show()
```
