6. Write a function definition named mode that takes in sequence of numbers and returns the most commonly occuring value

```python
def mode(numbers):
    freq = {}
    for num in numbers:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    max_count = 0
    mode_value = None
    for num, count in freq.items():
        if count > max_count:
            max_count = count
            mode_value = num
    return mode_value
```

Exercise 2: Replace all column values which contain ?, n.a, or NaN with suitable values and print the updated dataset.:

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Automobile_data.csv")
df = df.replace(['?', 'n.a'], np.nan)
#essentially we just replace all the ?, n.a and NaN with NaN first, and then replace them all by the mean to avoid distorting the data
mean_price = df['price'].mean()
df['price'] = df['price'].fillna(mean_price)
print(df)
```

Exercise 1: Read Total profit of all months and show it using a line plot \\
Total profit data provided for each month. Generated line plot must include the following properties: â€“

X label name = Month Number \\
Y label name = Total profit \\




```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("company_sales_data.csv")
months = df["month_number"]
total_profit = df["total_profit"]
plt.plot(months, total_profit)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Line Plot")
plt.show()
```

Exercise : Read face cream and facewash product sales data and show it using the bar chart \\
The bar chart should display the number of units sold per month for each product. \\
Add a separate bar for each product in the same chart.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("company_sales_data.csv")
month_number = df["month_number"]
facecream_sales = df["facecream"]
facewash_sales  = df["facewash"]
width = 0.4
plt.bar(month_number - width/2, facecream_sales, width, label = "FACECREAM")
plt.bar(month_number + width/2, facewash_sales, width, label = "FACEWASH")
plt.xlabel("month number")
plt.ylabel("rate")
plt.xticks(month_number)
plt.legend()
plt.show()
```
