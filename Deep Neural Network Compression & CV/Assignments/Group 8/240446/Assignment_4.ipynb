To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def analyze_image(image_rgb, title="Image", threshold=THRESHOLD):

    #convert RGB to grayscale
    #sobel works only on intensity changes
    gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

    # defining sobel kernels
    sobel_x = np.array([
        [-1,  0,  1],
        [-2,  0,  2],
        [-1,  0,  1]
    ])

    sobel_y = np.array([
        [-1, -2, -1],
        [ 0,  0,  0],
        [ 1,  2,  1]
    ])

    # convolution with sobel kernels
    Ix = cv2.filter2D(gray, cv2.CV_64F, sobel_x)
    Iy = cv2.filter2D(gray, cv2.CV_64F, sobel_y)

    # flatten gradients for scatter plot (converts into a 1D list for plotting as a scatter plot)
    Ix_flat = Ix.flatten()
    Iy_flat = Iy.flatten()

    # apply threshold
    mask = (np.abs(Ix_flat) > threshold) | (np.abs(Iy_flat) > threshold)
    Ix_thresh = Ix_flat[mask]
    Iy_thresh = Iy_flat[mask]

    # plot everything
    plt.figure(figsize=(10, 6))
    # original grayscale image
    plt.subplot(2, 2, 1)
    plt.imshow(gray, cmap='gray')
    plt.title(f"{title} (Grayscale)")
    plt.axis('off')
    # Ix
    plt.subplot(2, 2, 2)
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix")
    plt.axis('off')
    # Iy
    plt.subplot(2, 2, 3)
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy")
    plt.axis('off')
    # Iy vs Ix
    plt.subplot(2, 2, 4)
    plt.scatter(Ix_thresh, Iy_thresh, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(f"Iy vs Ix (Threshold = {threshold})")
    plt.show()

    return Ix, Iy
````

````python
#computing gradients first
def compute_gradients(gray):
    #sobel kernels
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]])

    sobel_y = np.array([[-1, -2, -1],
                        [ 0,  0,  0],
                        [ 1,  2,  1]])

    Ix = cv2.filter2D(gray, cv2.CV_64F, sobel_x)
    Iy = cv2.filter2D(gray, cv2.CV_64F, sobel_y)

    return Ix, Iy
````

````python
# now computing R map using second moments
def compute_R_map(image_rgb, window_size=5, k=0.04):
    #convert to grayscale
    gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

    #get Ix and Iy gradients using sobel
    Ix, Iy = compute_gradients(gray) #using our defined function

    #padding is to decide how much the window extends around each pixel
    pad = window_size // 2
    h, w = gray.shape
    R = np.zeros((h, w)) #stores the value for each pixel

    # loop over image pixels
    for y in range(pad, h - pad):
        for x in range(pad, w - pad):

            # take a small window around each pixel
            #analyze this local neighbourhood
            Ix_win = Ix[y-pad:y+pad+1, x-pad:x+pad+1]
            Iy_win = Iy[y-pad:y+pad+1, x-pad:x+pad+1]

            # second moments (a, b, c)
            a = np.sum(Ix_win ** 2)
            b = np.sum(Ix_win * Iy_win)
            c = np.sum(Iy_win ** 2)

            # ellipse response
            R[y, x] = (a * c - b ** 2) - k * (a + c) ** 2

    return R
````

````python
# Compute R-maps for the three given examples
R_flat   = compute_R_map(flat,   window_size=5, k=0.04)
R_edge   = compute_R_map(edge,   window_size=5, k=0.04)
R_corner = compute_R_map(corner, window_size=5, k=0.04)

#plot the R maps
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(R_flat, cmap='gray')
plt.title("Flat Region R-map")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(R_edge, cmap='gray')
plt.title("Edge Region R-map")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(R_corner, cmap='gray')
plt.title("Corner Region R-map")
plt.axis("off")
plt.show()
# corners are clearly visible in white
````

````python
# slope-intercept method
# function for slope-intercept hough
def hough_slope_intercept(image_rgb,
                          m_min=-5, m_max=5, m_steps=400,
                          c_min=-500, c_max=500, c_steps=400,
                          canny_low=100, canny_high=200,
                          acc_threshold_ratio=0.9):

    # convert image to grayscale since edges depend on intensity and not color
    gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

    # detect edges using Canny
    edges = cv2.Canny(gray, canny_low, canny_high)

    # create possible slope and intercept values
    m_vals = np.linspace(m_min, m_max, m_steps)
    c_vals = np.linspace(c_min, c_max, c_steps)

    # accumulator stores votes for each (m, c) pair
    accumulator = np.zeros((len(m_vals), len(c_vals)))

    # get coordinates of all edge pixels
    ys, xs = np.nonzero(edges)

    # for every edge pixel, vote for all possible lines passing through it
    for x, y in zip(xs, ys):
        for i, m in enumerate(m_vals):
            c = y - m * x  # line equation y = mx + c

            # ignore c values that fall outside the allowed range
            if c_min <= c <= c_max:
                j = int((c - c_min) / (c_max - c_min) * (c_steps - 1))
                accumulator[i, j] += 1

    # normalize accumulator so values lie between 0 and 1
    accumulator_norm = accumulator / accumulator.max()

    # pick only strong peaks from accumulator
    lines = np.argwhere(accumulator_norm > acc_threshold_ratio)

    # draw the detected lines on the original image
    output = image_rgb.copy()
    for i, j in lines:
        m = m_vals[i]
        c = c_vals[j]
        x0, x1 = 0, output.shape[1]
        y0, y1 = int(c), int(m * x1 + c)
        cv2.line(output, (x0, y0), (x1, y1), (0, 0, 255), 1)

    return edges, accumulator_norm, output
````

````python
edges_m, acc_m, out_m = hough_slope_intercept(img)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.imshow(edges_m, cmap='gray')
plt.title("Canny Edge Map")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(acc_m, cmap='hot')
plt.title("Accumulator (m-c)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(out_m)
plt.title("Detected Lines (m-c)")
plt.axis("off")

plt.show()
#chessboard as an image was a very bad choice for intercept-slope hough (due to many vertical lines-since slope tends to infinity)
#now we'll see how rho-theta hough is a much better and faster choice
````

````python
# rho-theta method
# function for rho-theta hough
def hough_rho_theta(image_rgb,
                    theta_steps=180,
                    rho_steps=300,
                    canny_low=100, canny_high=200,
                    acc_threshold_ratio=0.9):

    #convert to grayscale
    gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

    #detect edges first
    edges = cv2.Canny(gray, canny_low, canny_high)

    h, w = gray.shape
    diag_len = int(np.sqrt(h*h + w*w))  # max possible rho value

    #theta ranges from 0 to 180 degrees
    thetas = np.linspace(0, np.pi, theta_steps)
    rhos = np.linspace(-diag_len, diag_len, rho_steps)

    #accumulator for rho-theta space
    accumulator = np.zeros((len(rhos), len(thetas)))

    ys, xs = np.nonzero(edges)

    #voting in rho-theta space
    for x, y in zip(xs, ys):
        for t_idx, theta in enumerate(thetas):
            rho = x*np.cos(theta) + y*np.sin(theta)
            r_idx = np.argmin(np.abs(rhos - rho))
            accumulator[r_idx, t_idx] += 1

    #normalize accumulator
    acc_norm = accumulator / accumulator.max()

    #keep only strong peaks
    peaks = np.argwhere(acc_norm > acc_threshold_ratio)

    # draw detected lines
    output = image_rgb.copy()
    for r_idx, t_idx in peaks:
        rho = rhos[r_idx]
        theta = thetas[t_idx]

        #convert polar form back to line form
        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a*rho, b*rho
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))

        cv2.line(output, pt1, pt2, (0, 255, 0), 2)

    return edges, acc_norm, output
````

````python
edges_rt, acc_rt, out_rt = hough_rho_theta(img)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.imshow(edges_rt, cmap='gray')
plt.title("Canny Edge Map")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(acc_rt, cmap='hot')
plt.title("Accumulator (-)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(out_rt)
plt.title("Detected Lines (-)")
plt.axis("off")

plt.show()
#much cleaner and accurate image using rho-theta hough
````
