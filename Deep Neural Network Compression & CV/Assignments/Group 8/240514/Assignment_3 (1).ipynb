````python
def convolve(img, kernel):

    img = img.astype(np.float32)
    kernel = kernel.astype(np.float32)

    kh, kw = kernel.shape
    pad_h = kh // 2
    pad_w = kw // 2


    padded_img = np.pad(img,
                         ((pad_h, pad_h), (pad_w, pad_w)),
                         mode='reflect')

    h, w = img.shape
    output = np.zeros((h, w), dtype=np.float32)


    for i in range(h):
        for j in range(w):
            region = padded_img[i:i+kh, j:j+kw]
            output[i, j] = np.sum(region * kernel)

    return output
````

````python
def average_blur_gray(img, ksize):
    kernel = np.ones((ksize, ksize), dtype=np.float32)
    kernel /= (ksize * ksize)
    return convolve(img, kernel)
def average_blur_rgb(img, ksize):
    blurred = np.zeros_like(img, dtype=np.float32)

    for c in range(3):  # R, G, B
        blurred[:, :, c] = average_blur_gray(img[:, :, c], ksize)

    return blurred
````

````python
def gaussian_kernel(ksize, sigma):
    ax = np.arange(-(ksize // 2), ksize // 2 + 1)
    xx, yy = np.meshgrid(ax, ax)

    kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    kernel /= np.sum(kernel)  # normalize

    return kernel
def gaussian_blur_gray(img, ksize, sigma):
    kernel = gaussian_kernel(ksize, sigma)
    return convolve(img, kernel)
def gaussian_blur_rgb(img, ksize, sigma):
    blurred = np.zeros_like(img, dtype=np.float32)
    kernel = gaussian_kernel(ksize, sigma)

    for c in range(3):
        blurred[:, :, c] = convolve(img[:, :, c], kernel)

    return blurred
````

````python
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

sobel_y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
], dtype=np.float32)
Gx = convolve(gray_img, sobel_x)
Gy = convolve(gray_img, sobel_y)
gradient_magnitude = np.sqrt(Gx**2 + Gy**2)
def normalize(img):
    img_min = img.min()
    img_max = img.max()
    return ((img - img_min) / (img_max - img_min) * 255).astype(np.uint8)

Gx_norm = normalize(np.abs(Gx))
Gy_norm = normalize(np.abs(Gy))
grad_norm = normalize(gradient_magnitude)

threshold = 100
edge_map = grad_norm.copy()
edge_map[edge_map < threshold] = 0
plt.figure(figsize=(12, 4))

plt.plot(1, 4, 1)
plt.title("Gx")
plt.imshow(Gx_norm, cmap='gray')
plt.axis("off")
plt.show()
````

#### Task - Frequency Analysis (LPF & HPF Verification)
Compute the magnitude spectrum of the original image, gaussian blurred image and the unsharp masking image. Now, play with the alpha value in sharpening and gaussian kernel and notice how the magntude spectrum changes.
Then, in code comments, explain your observations of dark and bright areas in the magnitude spectrum. In nutshell, verify how convolution is actually filtering frequencies depending on the kernel.

````python
# Your PART D codes follow here. You can create more cells under this. Write clean and commented code.
````

````python
import numpy as np
import matplotlib.pyplot as plt
def magnitude_spectrum(img):

    f = np.fft.fft2(img)

    fshift = np.fft.fftshift(f)

    magnitude = 20 * np.log(np.abs(fshift) + 1)
    return magnitude
mag_original = magnitude_spectrum(gray_image)
mag_blurred = magnitude_spectrum(blurred)
mag_unsharp = magnitude_spectrum(sharp_unsharp)
````

````python
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
plt.imshow(mag_original, cmap='gray')
plt.title('Original Image Spectrum')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(mag_blurred, cmap='gray')
plt.title('Gaussian Blurred Spectrum (LPF)')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(mag_unsharp, cmap='gray')
plt.title('Unsharp Masking Spectrum (HPF)')
plt.axis('off')

plt.show()
````

````python
# Observations:
# 1. Dark areas in the magnitude spectrum represent low-frequency components
#    (smooth regions of the image, like background).
# 2. Bright areas represent high-frequency components (edges, fine details).
# 3. Gaussian blur attenuates high frequencies  spectrum becomes darker in high-frequency areas (LPF effect).
# 4. Unsharp masking enhances high frequencies  edges become brighter in the spectrum (HPF effect).
# 5. Convolution in spatial domain corresponds to multiplication in frequency domain:
#    - Blurring kernel (LPF) removes high-frequency components.
#    - Sharpening kernel (HPF) amplifies high-frequency components.
````
