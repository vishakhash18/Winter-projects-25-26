````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gray_histogram(gray_img):
  gray_img = np.array(gray_img)
  h, w = gray_img.shape

  hist = np.zeros(256, dtype = int)

  for i in range(h):
    for j in range(w):
      hist[gray_img[i,j]] += 1

  return hist

img = np.array(img)
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
gray = 0.299*R + 0.587*G + 0.114*B
gray = gray.astype(np.uint8)

plt.bar(np.arange(256), gray_histogram(gray), width = 1.0)
plt.title("Grayscale Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Total Pixels")
plt.show
````

Write a function <code>rgb_to_hsv</code> that takes a 0-255 ( not normalised ) RGB image and returns an HSV image ( also not normalised ).

````python
def rgb_to_hsv(image):
  image_norm = image.astype("float32") / 255.0
  R_norm = image_norm[:, :, 0]
  G_norm = image_norm[:, :, 1]
  B_norm = image_norm[:, :, 2]

  Cmax = np.max(image_norm, axis = 2)
  Cmin = np.min(image_norm, axis = 2)

  delta = Cmax - Cmin

  H = np.zeros_like(Cmax)

  mask = delta != 0            #(all the pixels where delta != 0) this will avoid division by 0
  r_mask = (Cmax == R_norm) & mask  #all pixels where Cmax == R and delta != 0 (binary and)
  g_mask = (Cmax == G_norm) & mask
  b_mask = (Cmax == B_norm) & mask

  H[r_mask] = (60 * (G_norm[r_mask] - B_norm[r_mask]) / delta[r_mask] + 360) % 360
  H[g_mask] = (60 * (B_norm[g_mask] - R_norm[g_mask]) / delta[g_mask] + 120) %360
  H[b_mask] = (60 * (R_norm[b_mask] - G_norm[b_mask]) / delta[b_mask] + 240) %360

  S = np.zeros_like(Cmax)
  nonzero = Cmax != 0          #this will also avoid division by 0
  S[nonzero] = delta[nonzero] / Cmax[nonzero]

  V = Cmax

  hsv = np.stack([H,S,V], axis = 2)
  return hsv

def hsv_to_rgb_cv2(hsv_image):
  H = hsv_image[:, :, 0]   # 0360
  S = hsv_image[:, :, 1]   # 01
  V = hsv_image[:, :, 2]   # 01

  H_cv = (H/2).astype(np.uint8)   #0-180
  S_cv = (S * 255).astype(np.uint8)   #0-255
  V_cv = (V * 255).astype(np.uint8)   #0-255

  hsv_cv = np.stack([H_cv, S_cv, V_cv], axis = 2)

  #now converting to rgb
  rgb_new = cv2.cvtColor(hsv_cv, cv2.COLOR_HSV2RGB)
  return rgb_new

hsv_image = rgb_to_hsv(img)
rgb_reconstructed = hsv_to_rgb_cv2(hsv_image)

# plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original RGB Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(rgb_reconstructed)
plt.title("Reconstructed RGB Image")
plt.axis("off")
plt.show()
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def ensure_uint8(img):
    arr = np.asarray(img)
    return np.clip(arr, 0, 255).astype(np.uint8)

def to_float01(img):
    return ensure_uint8(img).astype(np.float32) / 255.0

def to_uint8(img_f):
    return ensure_uint8((np.clip(img_f, 0.0, 1.0) * 255).astype(np.uint8))

def safe_clip(a, a_min, a_max):
    return np.clip(a, a_min, a_max)

def adjust_contrast(image, factor=1.0):
    f = to_float01(image)
    out = (f - 0.5) * factor + 0.5
    return to_uint8(out)

def change_saturation(image, scale=1.0):
    hsv = cv2.cvtColor(ensure_uint8(image), cv2.COLOR_RGB2HSV).astype(np.float32)
    h, s, v = cv2.split(hsv)
    s = safe_clip(s * scale, 0, 255)
    hsv2 = cv2.merge([h.astype(np.uint8), s.astype(np.uint8), v.astype(np.uint8)])
    return cv2.cvtColor(hsv2, cv2.COLOR_HSV2RGB)

def vibrance_boost(image, strength=0.3):
    hsv = cv2.cvtColor(ensure_uint8(image), cv2.COLOR_RGB2HSV).astype(np.float32)
    H, S, V = cv2.split(hsv)
    S_norm = S / 255.0
    S_new = (S_norm + strength * (1 - S_norm)) * 255.0
    hsv2 = cv2.merge([H.astype(np.uint8), S_new.astype(np.uint8), V.astype(np.uint8)])
    return cv2.cvtColor(hsv2, cv2.COLOR_HSV2RGB)

def gamma_correction(image, gamma=1.0):
    if gamma == 0:
        return ensure_uint8(image)
    f = to_float01(image)
    out = np.power(f, 1.0 / gamma)
    return to_uint8(out)

def drugaddict_filter(image):
    x = adjust_contrast(image, 1.18)
    x = change_saturation(x, 1.35)
    x = vibrance_boost(x, 0.55)
    return ensure_uint8(x)

def horror_filter(image):
    x = adjust_contrast(image, 0.90)
    x = gamma_correction(x, 1.12)
    x = change_saturation(x, 0.52)
    return ensure_uint8(x)

# The drugaddict_filter makes the image look harsh and worn-out by slightly increasing contrast and boosting both saturation and vibrance, which gives a rough, stressed, grungy appearance.

# The horror_filter makes the image look cold and eerie by lowering contrast, brightening shadows through gamma, and reducing saturation, creating a washed-out, lifeless horror vibe.

out_drug = drugaddict_filter(img_2)
out_horror = horror_filter(img_2)

plt.subplot(1,3,1)
plt.imshow(img_2)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(out_drug)
plt.title("After filter DrugAddict")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(out_horror)
plt.title("After filter Horror")
plt.axis("off")

plt.tight_layout()
plt.show()
````

#### White Balance Function
Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image. Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
# Your PART D codes follow here. You can create more cells under this. Write clean and commented code.
````

````python
def white_patch_balance(img):
  img = np.asarray(img)
  img_float = img.astype(np.float32)
  max_values = np.max(img_float, axis=(0,1))

  eps = 1e-6
  scale = 255.0 / (max_values + eps)

  balanced_img = img_float * scale.reshape(1,1,3)
  balanced_img = np.clip(balanced_img, 0, 255).astype(np.uint8)

  return balanced_img

balanced_img = white_patch_balance(img_2)

plt.subplot(1,2,1)
plt.imshow(img_2)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(balanced_img)
plt.title("White Balanced Image")
plt.axis("off")

plt.show()
````
