````python
def rgb_to_hsv(rgb_img):
  img=rgb_img.astype(np.float32)/255
  r=img[:,:,0]
  g=img[:,:,1]
  b=img[:,:,2]
  cmax=np.max(img, axis=2)
  cmin=np.min(img, axis=2)
  delta=cmax-cmin

  rows,cols=cmax.shape
  h=np.zeros((rows,cols))


  mask_r=(cmax==r)&(delta!=0)
  h[mask_r]=60*(((g[mask_r]-b[mask_r])/delta[mask_r])%6)

  mask_g=(cmax==g)&(delta!=0)
  h[mask_g]=60*(((b[mask_g]-r[mask_g])/delta[mask_g])+2)

  mask_b=(cmax==b)&(delta!=0)
  h[mask_b]=60*(((r[mask_b]-g[mask_b])/delta[mask_b])+4)

  s=np.zeros_like(cmax)

  mask_nz=(cmax != 0)
  s[mask_nz]=delta[mask_nz]/cmax[mask_nz]

  v=cmax

  h_cv=h/2

  s_cv= s*255

  v_cv=v*255

  hsv_img = np.stack([h_cv, s_cv, v_cv], axis=2)

  return np.round(hsv_img).astype(np.uint8)

my_hsv=rgb_to_hsv(img1_np)

reconverted_rgb = cv2.cvtColor(my_hsv, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(10,5))
plt.subplot(1, 2,1)
plt.imshow(img1_np)

plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(reconverted_rgb)
plt.title("Reconverted")
plt.axis('off')
plt.show()
````

````python
def cawfee_filter(img_rgb):


    img = img_rgb.astype(np.float32)

    alpha = 1.3
    beta = -20
    img = alpha * img + beta


    img = np.clip(img, 0, 255)


    r, g, b = cv2.split(img)
    r = r * 1.4
    b = b * 0.7


    merged = cv2.merge([r, g, b])
    return np.clip(merged, 0, 255).astype(np.uint8)



def shrek_ka_moot_wala_filter(img_rgb):


    gamma = 0.6


    lut = np.array([((i / 255.0) ** gamma) * 255
                    for i in range(256)]).astype("uint8")


    foggy_img = cv2.LUT(img_rgb, lut)

    foggy_float = foggy_img.astype(np.float32)
    r, g, b = cv2.split(foggy_float)

    g = g * 1.3
    r = r * 0.8


    merged = cv2.merge([r, g, b])

    return np.clip(merged, 0, 255).astype(np.uint8)

img_martian = cawfee_filter(img1_np)
img_swamp = shrek_ka_moot_wala_filter(img1_np)

# Display Results
plt.figure(figsize=(15, 6))

# Original
plt.subplot(1, 3, 1)
plt.imshow(img1_np)
plt.title("Original Image")
plt.axis('off')

# Filter 1
plt.subplot(1, 3, 2)
plt.imshow(img_martian)
plt.title("Cawfee Filter")
plt.axis('off')

# Filter 2
plt.subplot(1, 3, 3)
plt.imshow(img_swamp)
plt.title("Shrek Ka Moot Wala Filter")
plt.axis('off')

plt.tight_layout()
plt.show()
# I chose Cawfee Filter to enhance the scenery by increasing contrast and increasing  the color temperature of the image.
# I chose Shreak ka moot wala Filter to reduce the strong lighting while highlighting the greenery.
````

#### White Balance Function
Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image. Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
# Your PART D codes follow here. You can create more cells under this. Write clean and commented code.
````

````python
def white_patch_balance(img_rgb):

  img_float = img_rgb.astype(np.float32)

  max_vals = np.max(img_float, axis=(0, 1))

  scaling_factors = np.divide(255.0, max_vals, where=max_vals!=0)

  balanced_img = img_float * scaling_factors

  balanced_img = np.clip(balanced_img, 0, 255).astype(np.uint8)

  return balanced_img

img_out = white_patch_balance(img1_np)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img1_np)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_out)
plt.title("White Patch Balanced")
plt.axis('off')

plt.show()


````
