````python
import numpy as np
import cv2
import matplotlib.pyplot as plt
# size of image
rows,cols=gray_scale2.shape       #(rows,cols)
crow=rows//2      # row number of centred position of array
ccol=cols//2      # col number of centred position of array
# define function for LPF
plt.subplot(1,4,1)
plt.imshow(gray_scale2,cmap="gray")
plt.title("Original Image")
plt.axis("off")

def lpf(shape, radius):
    rows,cols=shape             # size of LPF
    crow=rows//2                # row number of centred position of LPF
    ccol=cols//2                # col number of centred position of LPF
    # setting each entry of unsigned 8 bit integer =0 in mask
    mask=np.zeros((rows,cols), dtype=np.uint8)
    #  creating a grid of cordinates
    y,x=np.ogrid[:rows, :cols]
    # calculate distance square of each pixel position from centred position
    distance_square=(x-ccol)**2 + (y-crow)**2
    # check points are inside a circle of radius centred at (crow,ccol)
    mask[distance_square <= radius**2] = 1  #set low frequencies position 1
    return mask
radii=[10,30,80]
   # filtered_images=[]   # empty list
for r in radii:
        mask=lpf(gray_scale2.shape,r)
        # setting frequencies out of circle of radius r to 0
        filtered_fshift=fshift*mask
        # getting back to image domain
        filtered_fshift_ishift=np.fft.ifftshift(filtered_fshift) # inverse shift
        filtered_image_complex=np.fft.ifft2(filtered_fshift_ishift)  # inverse fourier tranform to get complex image 2d array
        filtered_image=np.abs(filtered_image_complex) # interest in real part of pixels
      #  filtered_images.insert(filtered_image) # inserting in list
      # display images
        plt.subplot(1,4,radii.index(r)+2)
        plt.imshow(filtered_image,cmap="gray")
        plt.title("radius:"+str(r))
        plt.axis("off")



plt.show()
# as radius decreses very few low frequencies will stay in circle of particular radius or circle shrinks helps in removing edges make image more blur
# as radius become circle expands due to which high frequencies comes in the circle which makes the image less blur
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt
# High pass Filter
plt.subplot(1,4,1)
plt.imshow(gray_scale2,cmap="gray")
plt.title("Original Image")
plt.axis("off")
def hpf(shape,radius):
    mask=1-lpf(shape,radius)
    return mask
radii=[10,30,80]
for r in radii:
  mask=hpf(gray_scale2.shape,r)
  filtered_fshift=fshift*mask
  filtered_fshift_ishift=np.fft.ifftshift(filtered_fshift)
  filtered_image_complex=np.fft.ifft2(filtered_fshift_ishift)
  filtered_image=np.abs(filtered_image_complex)
  plt.subplot(1,4,radii.index(r)+2)
  plt.imshow(filtered_image,cmap="gray")
  plt.title("radius:"+str(r))
  plt.axis("off")





plt.show()
# high pass filter allows to keep high frequencies making low frequencies position 0 in fshift upto to a circle of particular radius
# low radius make a tiny spot at the centre where value is 0 and outside more high frequencies keep image more edged
# as radius increses ,spot at centre expands, it removes   larger low frequnecies keeping only very high frequencies making image more dark with thin edges
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

rows,cols=gray_scale2.shape
crow,ccol=rows//2,cols//2

def square_lpf(shape,side):
   rows,cols =shape
   crow,ccol=rows//2,cols//2
   mask=np.zeros((rows,cols),dtype=np.uint8)
   y,x=np.ogrid[:rows, :cols]
   mask[(y>=crow-side//2)&(y<=crow+side//2)&(x>=ccol-side//2)&(x<=ccol+side//2)]=1
   return mask

 # creating square low pass filter of side 50 of same size as grayscale image
mask=square_lpf(gray_scale2.shape,50) # side =50
#setting high frequencies out of square to 0
filtered_fshift=fshift*mask
#taking inverse shift for reconstruct
filtered_fshift_ishift=np.fft.ifftshift(filtered_fshift)
# taking inverse to get back in image domain (complex)
filtered_image_complex=np.fft.ifft2(filtered_fshift_ishift)  # 2d compex array
# taking only real values
filtered_image=np.abs(filtered_image_complex)


# display images
plt.subplot(1,3,1)
plt.imshow(gray_scale2,cmap="gray")
plt.title("original image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(filtered_image,cmap="gray")
plt.title("filtered image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(filtered_fshift.astype(np.uint8),cmap="gray")
plt.title("filtered_fshift")
plt.axis("off")

plt.show()
````
