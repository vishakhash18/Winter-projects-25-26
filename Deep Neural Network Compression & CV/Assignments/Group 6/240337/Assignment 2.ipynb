````python
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = """
"""  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
````

````python
img_rgb = np.array(img)

def gray_histogram(gray_img):
    histogram = np.zeros(256, dtype=int)

    height, width = gray_img.shape

    for i in range(height):
        for j in range(width):
            pixel_value = gray_img[i, j]
            histogram[pixel_value] += 1

    return histogram

gray_img = np.dot(img_rgb[...,:3], [0.299, 0.587, 0.114]).astype(np.uint8)

hist_gray = gray_histogram(gray_img)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.bar(range(256), hist_gray, color='black', width=1)
plt.title('Grayscale Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 255])
plt.tight_layout()
plt.show()

def rgb_histogram(rgb_img):
    """
    Compute histograms for each RGB channel using pure NumPy
    Input: rgb_img - RGB image with values 0-255
    Output: tuple of (hist_r, hist_g, hist_b)
    """
    hist_r = np.zeros(256, dtype=int)
    hist_g = np.zeros(256, dtype=int)
    hist_b = np.zeros(256, dtype=int)

    height, width, _ = rgb_img.shape

    for i in range(height):
        for j in range(width):
            hist_r[rgb_img[i, j, 0]] += 1
            hist_g[rgb_img[i, j, 1]] += 1
            hist_b[rgb_img[i, j, 2]] += 1

    return hist_r, hist_g, hist_b
hist_r, hist_g, hist_b = rgb_histogram(img_rgb)
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('RGB Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.plot(hist_r, color='red', alpha=0.7, label='Red')
plt.plot(hist_g, color='green', alpha=0.7, label='Green')
plt.plot(hist_b, color='blue', alpha=0.7, label='Blue')
plt.title('RGB Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend()
plt.xlim([0, 255])
plt.tight_layout()
plt.show()
````

````python
import cv2
def rgb_to_hsv(rgb_img):
    rgb_normalized = rgb_img.astype(np.float32) / 255.0

    r = rgb_normalized[:, :, 0]
    g = rgb_normalized[:, :, 1]
    b = rgb_normalized[:, :, 2]

    max_val = np.maximum(np.maximum(r, g), b)
    min_val = np.minimum(np.minimum(r, g), b)
    diff = max_val - min_val

    h = np.zeros_like(max_val)
    s = np.zeros_like(max_val)
    v = max_val

    non_zero_mask = max_val != 0
    s[non_zero_mask] = diff[non_zero_mask] / max_val[non_zero_mask]

    diff_non_zero = diff != 0

    r_max = (max_val == r) & diff_non_zero
    h[r_max] = (60 * ((g[r_max] - b[r_max]) / diff[r_max]) + 360) % 360

    g_max = (max_val == g) & diff_non_zero
    h[g_max] = (60 * ((b[g_max] - r[g_max]) / diff[g_max]) + 120) % 360

    b_max = (max_val == b) & diff_non_zero
    h[b_max] = (60 * ((r[b_max] - g[b_max]) / diff[b_max]) + 240) % 360

    hsv_img = np.zeros_like(rgb_img)
    hsv_img[:, :, 0] = (h / 2).astype(np.uint8)
    hsv_img[:, :, 1] = (s * 255).astype(np.uint8)
    hsv_img[:, :, 2] = (v * 255).astype(np.uint8)

    return hsv_img

hsv_custom = rgb_to_hsv(img_rgb)

rgb_reconstructed = cv2.cvtColor(hsv_custom, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Original RGB Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rgb_reconstructed)
plt.title('Reconstructed RGB (from custom HSV)')
plt.axis('off')
plt.tight_layout()
plt.show()

print("RGB to HSV conversion successful!")
print(f"HSV image shape: {hsv_custom.shape}")
print(f"H range: {hsv_custom[:,:,0].min()}-{hsv_custom[:,:,0].max()}")
print(f"S range: {hsv_custom[:,:,1].min()}-{hsv_custom[:,:,1].max()}")
print(f"V range: {hsv_custom[:,:,2].min()}-{hsv_custom[:,:,2].max()}")
````

````python
def sunset_glow_filter(rgb_img):
    img_float = rgb_img.astype(np.float32)

    img_float[:, :, 0] = np.clip(img_float[:, :, 0] * 1.15, 0, 255)
    img_float[:, :, 2] = np.clip(img_float[:, :, 2] * 0.85, 0, 255)

    hsv = cv2.cvtColor(img_float.astype(np.uint8), cv2.COLOR_RGB2HSV).astype(np.float32)
    hsv[:, :, 1] = np.clip(hsv[:, :, 1] * 1.3, 0, 255)
    img_float = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB).astype(np.float32)

    gamma = 1.2
    img_float = np.power(img_float / 255.0, 1/gamma) * 255.0

    contrast_factor = 1.1
    mean = img_float.mean()
    img_float = np.clip((img_float - mean) * contrast_factor + mean, 0, 255)

    return img_float.astype(np.uint8)


def moody_blue_filter(rgb_img):
    img_float = rgb_img.astype(np.float32)

    hsv = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2HSV).astype(np.float32)

    hsv[:, :, 1] = np.clip(hsv[:, :, 1] * 0.7, 0, 255)

    hsv[:, :, 2] = np.clip(hsv[:, :, 2] * 0.9, 0, 255)

    img_float = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB).astype(np.float32)

    luminance = 0.299 * img_float[:, :, 0] + 0.587 * img_float[:, :, 1] + 0.114 * img_float[:, :, 2]

    shadow_mask = (luminance < 100).astype(np.float32)
    shadow_mask = np.expand_dims(shadow_mask, axis=2)

    blue_tint = np.zeros_like(img_float)
    blue_tint[:, :, 2] = 30
    img_float = img_float + blue_tint * shadow_mask * 0.4

    contrast_factor = 1.25
    mean = img_float.mean()
    img_float = np.clip((img_float - mean) * contrast_factor + mean, 0, 255)

    hsv2 = cv2.cvtColor(img_float.astype(np.uint8), cv2.COLOR_RGB2HSV).astype(np.float32)
    sat_boost = (255 - hsv2[:, :, 1]) / 255.0 * 0.3
    hsv2[:, :, 1] = np.clip(hsv2[:, :, 1] + sat_boost * hsv2[:, :, 1], 0, 255)
    img_float = cv2.cvtColor(hsv2.astype(np.uint8), cv2.COLOR_HSV2RGB).astype(np.float32)

    return np.clip(img_float, 0, 255).astype(np.uint8)


filtered1 = sunset_glow_filter(img_rgb)
filtered2 = moody_blue_filter(img_rgb)

fig, axes = plt.subplots(2, 2, figsize=(14, 12))

axes[0, 0].imshow(img_rgb)
axes[0, 0].set_title('Original Image', fontsize=12, fontweight='bold')
axes[0, 0].axis('off')

axes[0, 1].imshow(filtered1)
axes[0, 1].set_title('Sunset Glow Filter', fontsize=12, fontweight='bold')
axes[0, 1].axis('off')

axes[1, 0].imshow(img_rgb)
axes[1, 0].set_title('Original Image', fontsize=12, fontweight='bold')
axes[1, 0].axis('off')

axes[1, 1].imshow(filtered2)
axes[1, 1].set_title('Moody Blue Dreams Filter', fontsize=12, fontweight='bold')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
````

#### White Balance Function
Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image. Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
# Your PART D codes follow here. You can create more cells under this. Write clean and commented code.
````

````python
def white_patch_balance(img_rgb):
    img_float = img_rgb.astype(np.float32)

    max_r = img_float[:, :, 0].max()
    max_g = img_float[:, :, 1].max()
    max_b = img_float[:, :, 2].max()

    scale_r = 255.0 / max_r if max_r > 0 else 1.0
    scale_g = 255.0 / max_g if max_g > 0 else 1.0
    scale_b = 255.0 / max_b if max_b > 0 else 1.0

    balanced = np.zeros_like(img_float)
    balanced[:, :, 0] = img_float[:, :, 0] * scale_r
    balanced[:, :, 1] = img_float[:, :, 1] * scale_g
    balanced[:, :, 2] = img_float[:, :, 2] * scale_b

    balanced = np.clip(balanced, 0, 255)

    return balanced.astype(np.uint8)

balanced_img = white_patch_balance(img_rgb)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(balanced_img)
plt.title('White Patch Balanced Image')
plt.axis('off')
plt.tight_layout()
plt.show()
````
