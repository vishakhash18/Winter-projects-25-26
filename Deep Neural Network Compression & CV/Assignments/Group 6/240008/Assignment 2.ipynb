````python
import numpy as np
import matplotlib.pyplot as plt

def gray_histogram(gray_img):
    """
    Input: 2D numpy array with values 0..255 (uint8 recommended).
    Computes histogram using explicit Python loops (no OpenCV, no np.bincount).
    Returns: hist (length-256 numpy array).
    """
    gray = np.asarray(gray_img)
    if gray.ndim != 2:
        raise ValueError("gray_histogram expects a 2D grayscale image.")

    # Ensure 0-255 integer range
    if gray.dtype != np.uint8:
        if gray.max() <= 1.0:
            gray = (gray * 255).astype(np.uint8)
        else:
            gray = gray.astype(np.uint8)

    rows, cols = gray.shape
    hist = np.zeros(256, dtype=np.int64)

    # Pure Python loops over pixels
    for r in range(rows):
        row = gray[r]
        for c in range(cols):
            v = int(row[c])      # ensure python int for indexing
            hist[v] += 1

    # Plot histogram
    plt.figure(figsize=(10,4))
    plt.bar(np.arange(256), hist, width=1.0)
    plt.title("Grayscale histogram (0-255)")
    plt.xlabel("Intensity")
    plt.ylabel("Count")
    plt.xlim(0,255)
    plt.tight_layout()
    plt.show()

    return hist
hist = gray_histogram(gray_numpy)

print("Total pixels counted:", hist.sum())

````

````python
def rgb_histogram(img_rgb):
    """
    Input: HxWx3 array (uint8 or convertible). Computes R, G, B histograms with pure loops.
    Returns: (hist_r, hist_g, hist_b)
    """
    arr = np.asarray(img_rgb)
    if arr.ndim != 3 or arr.shape[2] < 3:
        raise ValueError("rgb_histogram expects an HxWx3 RGB image.")

    # ensure first 3 channels only, ensure uint8
    arr = arr[:, :, :3]
    if arr.dtype != np.uint8:
        if arr.max() <= 1.0:
            arr = (arr * 255).astype(np.uint8)
        else:
            arr = arr.astype(np.uint8)

    rows, cols, _ = arr.shape
    hist_r = np.zeros(256, dtype=np.int64)
    hist_g = np.zeros(256, dtype=np.int64)
    hist_b = np.zeros(256, dtype=np.int64)

    for r in range(rows):
        row = arr[r]
        for c in range(cols):
            pr, pg, pb = row[c]
            hist_r[int(pr)] += 1
            hist_g[int(pg)] += 1
            hist_b[int(pb)] += 1

    # Plot histogram
    x = np.arange(256)
    plt.figure(figsize=(10,4))
    plt.plot(x, hist_r, color='r', label='Red')
    plt.plot(x, hist_g, color='g', label='Green')
    plt.plot(x, hist_b, color='b', label='Blue')
    plt.title("RGB Histograms")
    plt.xlabel("Intensity")
    plt.ylabel("Count")
    plt.xlim(0,255)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return hist_r, hist_g, hist_b

# call
hist_r, hist_g, hist_b = rgb_histogram(img_rgb)
print("R+G+B pixels sum checks:", hist_r.sum(), hist_g.sum(), hist_b.sum())
````

````python
# Your PART B codes follow here. You can create more cells under this. Write clean and commented code.
# Paste this whole block and run
import numpy as np
import matplotlib.pyplot as plt
import cv2            # make sure OpenCV is installed
from PIL import Image
import matplotlib.image as mpimg

def rgb_to_hsv_opencv_scale(img_rgb):
    """
    Input: img_rgb (HxWx3), dtype uint8 or convertible, values 0..255 (NOT normalized).
    Output: hsv (HxWx3) dtype uint8 with OpenCV scaling:
            H in 0..179, S in 0..255, V in 0..255
    Pure numpy implementation (vectorized).
    """
    # ensure numpy array and float computations
    rgb = np.asarray(img_rgb).astype(np.float32)

    if rgb.ndim != 3 or rgb.shape[2] < 3:
        raise ValueError("rgb_to_hsv_opencv_scale expects HxWx3 RGB image.")

    # keep only first 3 channels if there is alpha
    rgb = rgb[:, :, :3]

    R = rgb[:, :, 0]
    G = rgb[:, :, 1]
    B = rgb[:, :, 2]

    Cmax = np.maximum(np.maximum(R, G), B)
    Cmin = np.minimum(np.minimum(R, G), B)
    Delta = Cmax - Cmin

    # Prepare H, S, V (float)
    H = np.zeros_like(Cmax, dtype=np.float32)
    S = np.zeros_like(Cmax, dtype=np.float32)
    V = Cmax.copy()  # V is in 0..255 already

    # Avoid division by zero where Delta == 0
    nonzero = Delta > 1e-6  # tiny epsilon

    # When Cmax == R
    mask_r = (Cmax == R) & nonzero
    # ((G - B) / Delta) % 6  â€” use numpy modulus carefully
    # compute raw hue in degrees 0..360
    H_r = 60.0 * (( (G - B) / Delta ) % 6)
    H[mask_r] = H_r[mask_r]

    # When Cmax == G
    mask_g = (Cmax == G) & nonzero
    H_g = 60.0 * ( ( (B - R) / Delta ) + 2.0 )
    H[mask_g] = H_g[mask_g]

    # When Cmax == B
    mask_b = (Cmax == B) & nonzero
    H_b = 60.0 * ( ( (R - G) / Delta ) + 4.0 )
    H[mask_b] = H_b[mask_b]

    # For Delta == 0, H remains 0 (as desired)

    # Scale H from degrees (0..360) to OpenCV's 0..179
    H = (H / 2.0)

    # Saturation: S = (Delta / Cmax) * 255   (but if Cmax==0 => S=0)
    cmax_nonzero = Cmax > 1e-6
    S[cmax_nonzero] = (Delta[cmax_nonzero] / Cmax[cmax_nonzero]) * 255.0
    S[~cmax_nonzero] = 0.0

    # Clip to safe ranges (float) then convert to uint8
    H = np.clip(H, 0, 179)
    S = np.clip(S, 0, 255)
    V = np.clip(V, 0, 255)

    hsv = np.stack([H, S, V], axis=2).astype(np.uint8)
    return hsv

# -------------------------
# Example usage & debug
# -------------------------
# If you already have img_rgb from your base64 or file, use that.
# Otherwise we try to load 'my_image.jpeg' if present.
try:
    # Prefer existing variable if present in notebook
    img_rgb
except NameError:
    # fallback: try reading file my_image.jpeg in working directory
    try:
        img_rgb = mpimg.imread("my_image.jpeg")
        # mpimg may return floats 0..1, convert if needed
        if img_rgb.dtype != np.uint8:
            if img_rgb.max() <= 1.0:
                img_rgb = (img_rgb * 255).astype(np.uint8)
            else:
                img_rgb = img_rgb.astype(np.uint8)
    except Exception as e:
        raise RuntimeError("No img_rgb found and failed to load 'my_image.jpeg': " + str(e))

print("Input image shape, dtype, min/max:", img_rgb.shape, img_rgb.dtype, img_rgb.min(), img_rgb.max())

# Convert
hsv_manual = rgb_to_hsv_opencv_scale(img_rgb)
print("HSV produced shape/dtype/min/max per channel:")
print(" shape:", hsv_manual.shape, " dtype:", hsv_manual.dtype)
print(" H min,max:", hsv_manual[:,:,0].min(), hsv_manual[:,:,0].max())
print(" S min,max:", hsv_manual[:,:,1].min(), hsv_manual[:,:,1].max())
print(" V min,max:", hsv_manual[:,:,2].min(), hsv_manual[:,:,2].max())

# Convert back using OpenCV (HSV->RGB)
rgb_back = cv2.cvtColor(hsv_manual, cv2.COLOR_HSV2RGB)

print("Reconverted RGB shape/dtype/min/max:", rgb_back.shape, rgb_back.dtype, rgb_back.min(), rgb_back.max())

# Display side-by-side original and reconverted
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original RGB")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(rgb_back)
plt.title("RGB after (NumPy HSV -> OpenCV reconversion)")
plt.axis("off")

plt.show()
````

````python
# Your PART C codes follow here. You can create more cells under this. Write clean and commented code.
import numpy as np
import cv2
import matplotlib.pyplot as plt

# FILTER 1 : "Sunset-on-Steroids"
def sunset_on_steroids(img):
    """
    Input: 0â€“255 RGB image
    Output: 0â€“255 RGB image with warm, saturated sunset styling
    """

    rgb = img.astype(np.float32)

    # 1. Brightness lift
    rgb = rgb + 20  # mild brightness boost

    # 2. Warm tint (push reds & yellows)
    warm_matrix = np.array([1.10, 1.05, 0.95])  # Râ†‘, Gâ†‘, Bâ†“ slightly
    rgb = rgb * warm_matrix

    # 3. Saturation boost (using HSV)
    hsv = cv2.cvtColor(rgb.astype(np.uint8), cv2.COLOR_RGB2HSV)
    hsv[...,1] = np.clip(hsv[...,1] * 1.25, 0, 255)  # 25% more saturation
    rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB).astype(np.float32)

    # 4. Mild contrast stretch
    rgb = 1.05 * (rgb - 128) + 128

    # Final clipping
    return np.clip(rgb, 0, 255).astype(np.uint8)
````

````python
# FILTER 2 : "Frostbite Deluxeâ„¢"
def frostbite_deluxe(img):
    """
    Input: 0â€“255 RGB image
    Output: 0â€“255 cold, foggy, cinematic RGB image
    """

    rgb = img.astype(np.float32)

    #  1. Blue-green tint
    cold_matrix = np.array([0.95, 1.05, 1.12])  # Râ†“, Gâ†‘, Bâ†‘
    rgb = rgb * cold_matrix

    #  2. Gamma correction (cool, dark cinematic look)
    gamma = 1.2  # >1 means darker shadows
    rgb = 255 * ((rgb/255) ** gamma)

    #  3. Add fog layer (lift mid-tones)
    fog_strength = 25
    rgb = rgb + fog_strength

    #  4. Reduce contrast for dreamy look
    return np.clip(rgb, 0, 255).astype(np.uint8)
````

````python
import numpy as np
import matplotlib.pyplot as plt

def white_patch_balance(img_rgb):
    """
    img_rgb: HxWx3, uint8 (0-255)
    returns: white-balanced RGB image (uint8)
    """

    img = img_rgb.astype(np.float32)

    # Find channel-wise maxima
    max_r = img[:, :, 0].max()
    max_g = img[:, :, 1].max()
    max_b = img[:, :, 2].max()

    # Prevent division by zero
    max_r = max(max_r, 1.0)
    max_g = max(max_g, 1.0)
    max_b = max(max_b, 1.0)

    # Compute scale factors
    scale_r = 255.0 / max_r
    scale_g = 255.0 / max_g
    scale_b = 255.0 / max_b

    scales = np.array([scale_r, scale_g, scale_b])

    # Apply scaling
    wb = img * scales

    # Clip and cast to uint8
    wb = np.clip(wb, 0, 255).astype(np.uint8)

    return wb

````
