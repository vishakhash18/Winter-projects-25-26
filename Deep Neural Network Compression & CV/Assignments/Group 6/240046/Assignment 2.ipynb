````python
weights = np.array([0.299, 0.587, 0.114], dtype=np.float32)  # R, G, B
gray_numpy = (img_rgb.astype(np.float32) * weights.reshape((1,1,3))).sum(axis=2)
gray_numpy = np.clip(np.round(gray_numpy), 0, 255).astype(np.uint8)
````

````python
# Grayscale Histogram Function
def gray_histogram(gray_img):
    hist = np.zeros(256, dtype=int)
    H, W = gray_img.shape

    # pure loops, no numpy tricks
    for i in range(H):
        for j in range(W):
            val = gray_img[i, j]
            hist[val] += 1

    return hist


# Example Usage
hist_gray = gray_histogram(gray_img=gray_numpy)

plt.figure(figsize=(8,4))
plt.bar(np.arange(256), hist_gray, width=1.0, color='black')
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()


plt.subplot(1,3,3)
plt.imshow(gray_numpy, cmap='gray')
plt.axis('off')
plt.title('Gray Scale Image')

plt.subplot(1,3,2)
plt.imshow(img_rgb)
plt.axis('off')
plt.title('RGB Image')

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# RGB Histogram Function
def rgb_histogram(rgb_img):
    hist_r = np.zeros(256, dtype=int)
    hist_g = np.zeros(256, dtype=int)
    hist_b = np.zeros(256, dtype=int)

    H, W, _ = rgb_img.shape

    for i in range(H):
        for j in range(W):
            r, g, b = rgb_img[i, j]
            hist_r[r] += 1
            hist_g[g] += 1
            hist_b[b] += 1

    return hist_r, hist_g, hist_b

hist_r, hist_g, hist_b = rgb_histogram(img_rgb)

plt.figure(figsize=(10,4))
plt.plot(hist_r, color='red', label='Red')
plt.plot(hist_g, color='green', label='Green')
plt.plot(hist_b, color='blue', label='Blue')
plt.title("RGB Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.legend()
plt.show()
````

````python
# Input: uint8 RGB image (0255)
# Output: uint8 HSV image in OpenCV format (H:0179, S:0255, V:0255)

def rgb_to_hsv(rgb):
    rgb = rgb.astype(np.float32)
    R = rgb[:,:,0]
    G = rgb[:,:,1]
    B = rgb[:,:,2]

    Cmax = np.max(rgb, axis=2)
    Cmin = np.min(rgb, axis=2)
    delta = Cmax - Cmin + 1e-12
    # Hue
    H = np.zeros_like(Cmax)

    mask = (Cmax == R)
    H[mask] = (60 * ((G-B)/delta))[mask]

    mask = (Cmax == G)
    H[mask] = (60 * ((B-R)/delta + 2))[mask]

    mask = (Cmax == B)
    H[mask] = (60 * ((R-G)/delta + 4))[mask]

    H = np.mod(H, 360)
    H = H * (179/360)
    H = H.astype(np.uint8)

    # Saturation
    S = (delta / (Cmax + 1e-12)) * 255
    S[Cmax == 0] = 0             # black pixels  saturation = 0
    S = S.astype(np.uint8)

    # Value
    V = Cmax.astype(np.uint8)

    hsv = np.dstack([H, S, V])
    return hsv
````

````python
# Filter 1: "TropicalBoost"
# Effects:
#  - boost overall contrast and midtone vibrance
#  - increase saturation (but protect highlights)
#  - subtle warm tint in highlights and cool-blue tint in shadows (split-toning)
#  - slight hue shift toward warm (small rotation)

#Good for:
# - Tropical landscapes, beaches, rainforests, waterfalls, and travel photos
# - Images with lots of greens/blues where a vibrant, lively color pop is desired
# - Scenes with foliage, oceans, and skies that benefit from warmer highlights,
#   deeper shadows, and enhanced midtone saturation

def tropical_boost(rgb_in):
    img = rgb_in.astype(np.float32)

    # 1) contrast & brightness (simple linear contrast with clipping)
    #    contrast factor >1 increases contrast, bias shifts midtones
    contrast = 1.12
    brightness_bias = 5.0
    img = np.clip((img - 128.0) * contrast + 128.0 + brightness_bias, 0, 255)

    # 2) mild gamma to pop midtones (gamma <1 brightens)
    gamma = 0.95
    img = 255.0 * ((img / 255.0) ** (1.0 / gamma))

    # 3) convert to HSV to modify saturation and hue
    hsv = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_RGB2HSV).astype(np.float32)
    H, S, V = cv2.split(hsv)

    # 3a) hue rotation: small positive shift to warm (OpenCV H range 0-179)
    H = (H + 6) % 180

    # 3b) saturation boost with highlight protection:
    # increase saturation more for midtones (vibrance effect)
    # compute vibrance mask from V: midtones have higher factor
    v_norm = V / 255.0
    vibrance_factor = 1.25
    vibrance_mask = 1.0 - np.abs(v_norm - 0.5) * 1.6  # peak at midtones
    vibrance_mask = np.clip(vibrance_mask, 0.0, 1.0)
    S = S * (1.0 + (vibrance_mask * (vibrance_factor - 1.0)))
    S = np.clip(S, 0, 255)

    # 4) split-toning: warm highlights, cool shadows using lightweight channel mixing in RGB domain
    hsv_merged = cv2.merge([H, S, V]).astype(np.uint8)
    img2 = cv2.cvtColor(hsv_merged, cv2.COLOR_HSV2RGB).astype(np.float32)

    # compute shadow/highlight masks from V
    shadow_mask = np.clip((128 - V) / 128.0, 0.0, 1.0)   # 1 for deep shadows
    highlight_mask = np.clip((V - 128) / 127.0, 0.0, 1.0) # 1 for bright highlights

    # apply cool-blue tint to shadows (add to B channel)
    img2[:,:,2] = img2[:,:,2] + 30.0 * shadow_mask  # boost blue in shadows

    # apply warm tint to highlights (increase R channel)
    img2[:,:,0] = img2[:,:,0] + 18.0 * highlight_mask  # boost red in highlights

    # 5) final gentle sharpen via unsharp mask (improves perceived detail)
    blurred = cv2.GaussianBlur(img2, (0,0), sigmaX=1.0)
    sharpen = img2 + 0.6 * (img2 - blurred)
    out = np.clip(sharpen, 0, 255).astype(np.uint8)

    return out
````

````python
# Filter 2: "CinematicSunset"
# Effects:
#  - Apply a soft S-curve contrast boost for cinematic depth
#  - Warm color temperature by increasing red and slightly lowering blue
#  - Boost vibrance (mid-saturation regions) while protecting highlights
#  - Split-toning: warm highlights + teal shadows (classic film look)
#  - Add subtle luminance-only film grain for texture
#  - Apply vignette to focus attention toward the center
#  - Final mild tone mapping for cohesive cinematic output

# Good for:
#  - Sunset photos, golden hour portraits, warm landscapes, beach scenes,
#    and any image where a dramatic cinematic color grade is desired.
def cinematic_sunset(img_rgb):
    img = img_rgb.astype(np.float32)

    # 1) Contrast & brightness (S-curve using simple piecewise sigmoid-like)
    mean = 128.0
    contrast = 1.15
    img = (img - mean) * contrast + mean
    img = np.clip(img, 0, 255)

    # 2) Tiny color temperature shift: boost red, slightly reduce blue
    img[:,:,0] = np.clip(img[:,:,0] * 1.06, 0, 255)  # R up
    img[:,:,2] = np.clip(img[:,:,2] * 0.96, 0, 255)  # B down

    # 3) Vibrance: increase saturation more for mid-saturation pixels (protect highlights)
    hsv = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_RGB2HSV).astype(np.float32)
    H, S, V = cv2.split(hsv)
    # vibrance factor curve (stronger for mid S)
    s_norm = S / 255.0
    vib_factor = 1.0 + 0.45 * (1.0 - np.abs(s_norm - 0.5) * 2.0)  # peak at mid S
    S = np.clip(S * vib_factor, 0, 255)

    # 4) Split-toning: warm highlights, subtle teal shadows
    hsv_back = cv2.merge([H, S, V]).astype(np.uint8)
    rgb2 = cv2.cvtColor(hsv_back, cv2.COLOR_HSV2RGB).astype(np.float32)
    v_norm = V / 255.0
    shadow_mask = np.clip((0.5 - v_norm) * 2.0, 0.0, 1.0)  # 1 for dark
    highlight_mask = np.clip((v_norm - 0.6) * 2.5, 0.0, 1.0)  # 1 for bright
    # apply teal to shadows (add to B and G), warm to highlights (add to R)
    rgb2[:,:,1] += 18.0 * shadow_mask   # G
    rgb2[:,:,2] += 30.0 * shadow_mask   # B
    rgb2[:,:,0] += 22.0 * highlight_mask  # R
    rgb2 = np.clip(rgb2, 0, 255)

    # 5) Film grain (subtle, luminance-preserving)
    Ht, Wt = rgb2.shape[:2]
    grain = (np.random.randn(Ht, Wt) * 7.0).astype(np.float32)  # std dev ~7
    # add grain to luminance channel
    lum = 0.2989*rgb2[:,:,0] + 0.5870*rgb2[:,:,1] + 0.1140*rgb2[:,:,2]
    lum = np.clip(lum + grain, 0, 255)
    # reconstruct color with original chroma scaled to new luminance
    eps = 1e-6
    chroma_r = rgb2[:,:,0] / ( (rgb2[:,:,0]+rgb2[:,:,1]+rgb2[:,:,2]) + eps )
    chroma_g = rgb2[:,:,1] / ( (rgb2[:,:,0]+rgb2[:,:,1]+rgb2[:,:,2]) + eps )
    chroma_b = rgb2[:,:,2] / ( (rgb2[:,:,0]+rgb2[:,:,1]+rgb2[:,:,2]) + eps )
    rgb3 = np.zeros_like(rgb2)
    rgb3[:,:,0] = lum * chroma_r
    rgb3[:,:,1] = lum * chroma_g
    rgb3[:,:,2] = lum * chroma_b
    rgb3 = np.clip(rgb3, 0, 255)

    # 6) Gentle vignette
    y = np.linspace(-1,1,Ht)[:,None]
    x = np.linspace(-1,1,Wt)[None,:]
    mask = np.sqrt(x*x + y*y)
    vignette = np.clip(1.0 - 0.55 * (mask**1.6), 0.0, 1.0)
    rgb3[:,:,0] *= vignette
    rgb3[:,:,1] *= vignette
    rgb3[:,:,2] *= vignette

    # 7) Final tone mapping (soft contrast) and uint8 conversion
    out = np.clip(rgb3 * 1.02 + 3.0, 0, 255).astype(np.uint8)
    return out
````

````python
out1 = tropical_boost(img_rgb)
out2 = cinematic_sunset(img_rgb)

plt.figure(figsize=(15,6))
plt.subplot(1,3,1); plt.imshow(img_rgb); plt.axis('off'); plt.title('Original')
plt.subplot(1,3,2); plt.imshow(out1); plt.axis('off'); plt.title('TropicalBoost')
plt.subplot(1,3,3); plt.imshow(out2); plt.axis('off'); plt.title('CinematicSunset')
plt.show()
````

#### White Balance Function
Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image. Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
# Your PART D codes follow here. You can create more cells under this. Write clean and commented code.
````

````python
def white_patch_balance(img_rgb):
    img = img_rgb.astype(np.float32)

    # brightest R, G, B in the image
    max_r = np.max(img[:,:,0])
    max_g = np.max(img[:,:,1])
    max_b = np.max(img[:,:,2])

    # scale factors so brightest becomes 255
    sr = 255.0 / (max_r + 1e-12)
    sg = 255.0 / (max_g + 1e-12)
    sb = 255.0 / (max_b + 1e-12)

    # apply scaling
    out = img.copy()
    out[:,:,0] = out[:,:,0] * sr
    out[:,:,1] = out[:,:,1] * sg
    out[:,:,2] = out[:,:,2] * sb

    # clip to valid range
    out = np.clip(out, 0, 255).astype(np.uint8)
    return out

balanced = white_patch_balance(img_rgb)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1); plt.imshow(img_rgb); plt.axis('off'); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(balanced); plt.axis('off'); plt.title("White Patch Balanced")
plt.show()
````

````python
def white_balance(img_rgb, value):
    img = img_rgb.astype(np.float32)

    # value: 0  cold, 0.5  original, 1  warm
    # map value into [-1, +1]
    t = (value - 0.5) * 2.0

    # strength curves
    # Red shifts positively  warmer
    # Blue shifts negatively  colder
    red_gain  = 1.0 + 0.20 * t        # +20% max warming
    blue_gain = 1.0 - 0.20 * t        # -20% when warm (blue reduced)
    green_gain = 1.0 + 0.06 * t       # small green compensation

    # apply channel-wise white balance
    out = img.copy()
    out[:,:,0] *= red_gain
    out[:,:,1] *= green_gain
    out[:,:,2] *= blue_gain

    # clip to valid range
    out = np.clip(out, 0, 255).astype(np.uint8)
    return out

cold  = white_balance(img_rgb, 0.0)
orig  = white_balance(img_rgb, 0.5)
warm  = white_balance(img_rgb, 1.0)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1); plt.imshow(cold); plt.axis('off'); plt.title("Value = 0 (Coldest)")
plt.subplot(1,3,2); plt.imshow(orig); plt.axis('off'); plt.title("Value = 0.5 (Original)")
plt.subplot(1,3,3); plt.imshow(warm); plt.axis('off'); plt.title("Value = 1 (Warmest)")
plt.show()
````
