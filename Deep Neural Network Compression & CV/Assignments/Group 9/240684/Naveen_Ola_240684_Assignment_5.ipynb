```python
def fit_line(X,y,lr,epochs,l2_lambda):
  m,n=X.shape
  weights=np.zeros(n)
  bias=0
  loss_history =[]
  for i in range(epochs):
    y_predict=np.dot(X,weights)+bias
    error=y_predict-y
    dw = (1 / m) * np.dot(X.T, error) + (l2_lambda * weights)
    db = (1 / m) * np.sum(error)
    weights -= lr * dw
    bias -= lr * db
    loss = (1 / (2 * m)) * np.sum(error**2) + (l2_lambda / 2) * np.sum(weights**2)
    loss_history.append(loss)

    return weights, bias, loss_history
```

```python
# solution to question 2
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def logistic_regression_fit(X, y, lr=0.01, epochs=1000, l2_lambda=0.0):
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0.0
    losses = []

    for _ in range(epochs):
        z = np.dot(X, w) + b
        y_hat = sigmoid(z)

        loss = (
            -np.mean(y * np.log(y_hat + 1e-9) + (1 - y) * np.log(1 - y_hat + 1e-9))
            + (l2_lambda / (2 * n_samples)) * np.sum(w ** 2)
        )
        losses.append(loss)

        dw = (1 / n_samples) * np.dot(X.T, (y_hat - y)) + (l2_lambda / n_samples) * w
        db = (1 / n_samples) * np.sum(y_hat - y)

        w -= lr * dw
        b -= lr * db

    return w, b, losses
```

```python
def logistic_predict_proba(X, w, b):
    return sigmoid(np.dot(X, w) + b)
```

```python
def logistic_predict(X, w, b):
    return (logistic_predict_proba(X, w, b) >= 0.5).astype(int)
```
