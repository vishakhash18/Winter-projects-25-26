To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# Convolution function that convolves an image with a given kernel
def convolve(img, kernel):
    img = np.array(img)
    kernel = np.array(kernel)

    kh, kw = kernel.shape
    ih, iw = img.shape

    # padding size
    pad_h = kh // 2
    pad_w = kw // 2

    # pad image using reflect mode
    padded_img = np.pad(img, ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')

    # output image
    output = np.zeros((ih, iw))

    # convolution operation
    for i in range(ih):
        for j in range(iw):
            region = padded_img[i:i+kh, j:j+kw]
            output[i, j] = np.sum(region * kernel) 

    return output
````

````python
#function that return Ix and Iy using sobel operator
import numpy as np
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

sobel_y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
])

def sobel_gradients(gray_img, threshold = None, check = 0): #create a check if check is 1 then plot Ix, Iy and scatter plot of Iy vs Ix
    Ix = convolve(gray_img, sobel_x)
    Iy = convolve(gray_img, sobel_y)

    if(check == 1):
        plt.figure(figsize=(12,4))
    
        plt.subplot(1,3,1)
        plt.imshow(Ix, cmap='gray')
        plt.title("Ix")
        plt.colorbar()
    
        plt.subplot(1,3,2)
        plt.imshow(Iy, cmap='gray')
        plt.title("Iy")
        plt.colorbar()
    
        plt.subplot(1,3,3)
        plt.scatter(Ix.flatten(), Iy.flatten(), s=10)
        plt.xlabel("Ix")
        plt.ylabel("Iy")
        plt.title("Iy vs Ix Scatter Plot")
    
        plt.tight_layout()
        plt.show()

    if threshold is not None:
        magnitude = np.sqrt(Ix**2 + Iy**2)
        edge = np.zeros_like(gray_img)
        edge[magnitude >= threshold] = 255

        return Ix, Iy, edge
    
    return Ix, Iy
````

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def R_map(gray, Ix, Iy, window_size=3, k=0.04):
    pad = window_size // 2
    h, w = gray.shape
    
    R_map = np.zeros((h, w))
    
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy
    
    Ix2_p = np.pad(Ix2, pad)
    Iy2_p = np.pad(Iy2, pad)
    Ixy_p = np.pad(Ixy, pad)
    
    for i in range(h):
        for j in range(w):
            a = np.sum(Ix2_p[i:i+window_size, j:j+window_size])
            b = np.sum(Ixy_p[i:i+window_size, j:j+window_size])
            c = np.sum(Iy2_p[i:i+window_size, j:j+window_size])
            
            det = a*c - b*b
            trace = a + c
            R_map[i, j] = det - k * (trace ** 2)
    
    return R_map
````

### Part C - Hough Transforms

Again, import an image, create a function that takes the image, finds the canny edge map, then loops through it to find the equivalent line of each pixel, checks whether the slope and intercept are within acceptable limits of the accumulator size ( if not, discard ), does the voting procedure and creates the accumulator. Now, choose an accurate threshold for the accumulator to get the slope and intercept for a good number of lines in the image. One thing you can do is normalise the accumulator and use a 90 percent threshold ( just an example, im not telling if 90p is a good one or not ), then, mark the lines in the image using cv2 and display the final output. For the slope and intercept, choose the limit as per your image to get a good output. Please dont make it super large so it take hours to run and give the output.

````python
import cv2
def hough_mc(gray):
    
    edges = cv2.Canny(gray, 100, 200) #canny edge detection using cv2, implemented using numpy also

    h, w = edges.shape

    #Now we will se the parameter space for m and c
    m_min, m_max = -2, 2          # slope range
    c_min, c_max = -w, w          # intercept range

    m_bins = 400
    c_bins = 400

    m_vals = np.linspace(m_min, m_max, m_bins)
    c_vals = np.linspace(c_min, c_max, c_bins)

    accumulator = np.zeros((m_bins, c_bins), dtype=np.uint64)

    # voting
    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x
            if c_min <= c <= c_max:
                ci = int((c - c_min) / (c_max - c_min) * (c_bins - 1))
                accumulator[mi, ci] += 1


    acc_norm = accumulator / accumulator.max()

    threshold = 0.9
    lines = np.argwhere(acc_norm > threshold)

    output = gray.copy()
    for mi, ci in lines:
        m = m_vals[mi]
        c = c_vals[ci]

        x1, x2 = 0, w
        y1 = int(m * x1 + c)
        y2 = int(m * x2 + c)

        if 0 <= y1 < h and 0 <= y2 < h:
            cv2.line(output, (x1, y1), (x2, y2), (0, 0, 255), 1)

    return edges, accumulator, output


edges, acc, result = hough_mc(corner_gray)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1); plt.title("Edges"); plt.imshow(edges, cmap='gray')
plt.subplot(1,3,2); plt.title("Accumulator"); plt.imshow(acc, cmap='hot')
plt.subplot(1,3,3); plt.title("Detected Lines"); plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
````

````python
def hough_rho_theta(gray):

    edges = cv2.Canny(gray, 100, 200)

    h, w = edges.shape
    diag_len = int(np.sqrt(h*h + w*w))

    rhos = np.linspace(-diag_len, diag_len, 400)
    thetas = np.deg2rad(np.linspace(-90, 90, 180))

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.uint64)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for ti, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            ri = int((rho + diag_len) / (2 * diag_len) * (len(rhos) - 1))
            accumulator[ri, ti] += 1

    acc_norm = accumulator / accumulator.max()

    threshold = 0.9
    lines = np.argwhere(acc_norm > threshold)

    output = gray.copy()
    for ri, ti in lines:
        rho = rhos[ri]
        theta = thetas[ti]

        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a * rho, b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(output, (x1, y1), (x2, y2), (0, 255, 0), 1)

    return edges, accumulator, output

edges, acc, result = hough_rho_theta(corner_gray)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1); plt.title("Edges"); plt.imshow(edges, cmap='gray')
plt.subplot(1,3,2); plt.title("Accumulator"); plt.imshow(acc, cmap='hot')
plt.subplot(1,3,3); plt.title("Detected Lines"); plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
````
