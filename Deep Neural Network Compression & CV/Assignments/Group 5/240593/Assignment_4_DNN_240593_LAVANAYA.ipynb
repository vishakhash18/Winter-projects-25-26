To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#we will create a function that does the sobel analysis
def sobel_analysis(gray_img, title="", threshold=None):
    """
    gray_img  : input grayscale image
    title     : title for plots
    threshold : if not None, gradients with magnitude < threshold are ignored

    returns: Ix, Iy
    """

    # Sobel gradients
    Ix = cv2.Sobel(gray_img, cv2.CV_64F, dx=1, dy=0, ksize=3)
    Iy = cv2.Sobel(gray_img, cv2.CV_64F, dx=0, dy=1, ksize=3)

    # Optional thresholding
    if threshold is not None:
        mag = np.sqrt(Ix**2 + Iy**2)
        mask = mag > threshold
        Ix = Ix[mask]
        Iy = Iy[mask]
    else:
        Ix = Ix.flatten()
        Iy = Iy.flatten()

    # Plot Ix and Iy images
    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.imshow(cv2.convertScaleAbs(cv2.Sobel(gray_img, cv2.CV_64F, 1, 0)), cmap='gray')
    plt.title(f"{title} - Ix")
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.imshow(cv2.convertScaleAbs(cv2.Sobel(gray_img, cv2.CV_64F, 0, 1)), cmap='gray')
    plt.title(f"{title} - Iy")
    plt.axis('off')

    # Scatter plot Iy vs Ix
    plt.subplot(1,3,3)
    plt.scatter(Ix, Iy, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(f"{title} - Iy vs Ix")

    plt.tight_layout()
    plt.show()

    return Ix, Iy

````

````python
sobel_analysis(gray_flat,title="flat", threshold=None)
sobel_analysis(gray_edge,title="edge", threshold=None)
sobel_analysis(gray_corner,title="corner", threshold=None)
````

````python
def harris_r_map(gray_img, window_size=3, k=0.04):


    # Sobel gradients
    Ix = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=3)

    h, w = gray_img.shape
    offset = window_size // 2

    R_map = np.zeros((h, w))
    lambda1_map = np.zeros((h, w))
    lambda2_map = np.zeros((h, w))

    # Loop over pixels
    for y in range(offset, h - offset):
        for x in range(offset, w - offset):

            # Window
            Ix_w = Ix[y-offset:y+offset+1, x-offset:x+offset+1]
            Iy_w = Iy[y-offset:y+offset+1, x-offset:x+offset+1]

            # Second moments
            a = np.sum(Ix_w ** 2)
            b = np.sum(Iy_w ** 2)
            c = np.sum(Ix_w * Iy_w)

            # Eigenvalues
            trace = a + b
            det = a * b - c ** 2
            temp = np.sqrt((a - b) ** 2 + 4 * c ** 2)

            lambda1 = (trace + temp) / 2
            lambda2 = (trace - temp) / 2

            # Harris response
            R = det - k * (trace ** 2)

            R_map[y, x] = R
            lambda1_map[y, x] = lambda1
            lambda2_map[y, x] = lambda2

    return R_map, lambda1_map, lambda2_map
````

````python
def show_r_map(R_map, title="R Map"):
    R_norm = cv2.normalize(R_map, None, 0, 255, cv2.NORM_MINMAX)
    R_norm = R_norm.astype(np.uint8)

    plt.figure(figsize=(5,5))
    plt.imshow(R_norm, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
````

````python
#applpy this rmap to images
# Flat image
R_flat, l1_f, l2_f = harris_r_map(gray_flat)
show_r_map(R_flat, "Flat Image - R Map")

# Edge image
R_edge, l1_e, l2_e = harris_r_map(gray_edge)
show_r_map(R_edge, "Edge Image - R Map")

# Corner image
R_corner, l1_c, l2_c = harris_r_map(gray_corner)
show_r_map(R_corner, "Corner Image - R Map")

````

````python
def hough_mb(gray,
             m_min=-2, m_max=2, m_bins=200,
             b_min=-500, b_max=500, b_bins=200):

    edges = cv2.Canny(gray, 100, 200)

    h, w = gray.shape
    m_vals = np.linspace(m_min, m_max, m_bins)
    b_vals = np.linspace(b_min, b_max, b_bins)

    accumulator = np.zeros((m_bins, b_bins))

    # Loop through edge pixels
    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for i, m in enumerate(m_vals):
            b = y - m * x
            if b_min <= b <= b_max:
                j = int((b - b_min) / (b_max - b_min) * (b_bins - 1))
                accumulator[i, j] += 1

    return accumulator, m_vals, b_vals, edges
````

````python
acc, m_vals, b_vals, edges = hough_mb(gray)

# Normalize accumulator
acc_norm = acc / acc.max()

# Threshold (90%)
threshold = 0.9
lines = np.where(acc_norm > threshold)
````

````python
img_mb = img.copy()

for i, j in zip(lines[0], lines[1]):
    m = m_vals[i]
    b = b_vals[j]

    x1, x2 = 0, img.shape[1]
    y1 = int(m * x1 + b)
    y2 = int(m * x2 + b)

    if 0 <= y1 < img.shape[0] and 0 <= y2 < img.shape[0]:
        cv2.line(img_mb, (x1, y1), (x2, y2), (0,0,255), 1)

plt.imshow(cv2.cvtColor(img_mb, cv2.COLOR_BGR2RGB))
plt.title("Slope-Intercept Hough Lines")
plt.axis("off")
plt.show()
````

````python
def hough_rho_theta(gray):

    edges = cv2.Canny(gray, 100, 200)
    h, w = gray.shape

    thetas = np.deg2rad(np.arange(-90, 90))
    diag_len = int(np.sqrt(h*h + w*w))
    rhos = np.linspace(-diag_len, diag_len, 2*diag_len)

    accumulator = np.zeros((len(rhos), len(thetas)))

    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for t_idx, theta in enumerate(thetas):
            rho = int(x * np.cos(theta) + y * np.sin(theta)) + diag_len
            accumulator[rho, t_idx] += 1

    return accumulator, rhos, thetas
````

````python
acc_rt, rhos, thetas = hough_rho_theta(gray)

acc_norm = acc_rt / acc_rt.max()
threshold = 0.5   # lower works better here

lines = np.where(acc_norm > threshold)
````

````python
img_rt = img.copy()

for rho_idx, theta_idx in zip(lines[0], lines[1]):
    rho = rhos[rho_idx]
    theta = thetas[theta_idx]

    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho

    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))

    cv2.line(img_rt, (x1, y1), (x2, y2), (0,255,0), 1)

plt.imshow(cv2.cvtColor(img_rt, cv2.COLOR_BGR2RGB))
plt.title("Rho-Theta Hough Lines")
plt.axis("off")
plt.show()
````

````python
#Create a python function that will take the first image and a set of coordinates ( x1,y1 ) - top left and ( x2,y2 ) - bottom right to form a box around the white ball. Now, find gradients, select the biggest completed gradient ( maximum area contour ) and then create an RGBA ( 4 channel ) stencil.

def create_ball_rgba_stencil(img, x1, y1, x2, y2):
    """
    img : input BGR image
    (x1,y1) : top-left of bounding box
    (x2,y2) : bottom-right of bounding box

    returns: RGBA image (4-channel)
    """


    # 1. Crop the bounding box

    crop = img[y1:y2, x1:x2]

    gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)

    # 2. Compute gradients

    Ix = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

    grad_mag = np.sqrt(Ix**2 + Iy**2)
    grad_mag = cv2.normalize(grad_mag, None, 0, 255, cv2.NORM_MINMAX)
    grad_mag = grad_mag.astype(np.uint8)

    # 3. Edge detection

    edges = cv2.Canny(grad_mag, 50, 150)

    # 4. Find contours

    contours, _ = cv2.findContours(
        edges,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )

    if len(contours) == 0:
        raise ValueError("No contours found in the bounding box")

    # 5. Largest closed contour

    largest_contour = max(contours, key=cv2.contourArea)
-
    # 6. Create alpha mask

    alpha = np.zeros(gray.shape, dtype=np.uint8)
    cv2.drawContours(alpha, [largest_contour], -1, 255, thickness=-1)


    # 7. Create RGBA image

    b, g, r = cv2.split(crop)
    rgba = cv2.merge((b, g, r, alpha))

    return rgba

````

````python
# Bounding box around the ball
x1, y1 = 120, 80
x2, y2 = 320, 280

rgba_ball = create_ball_rgba_stencil(img, x1, y1, x2, y2)
````

````python
def overlay_stencil(background, rgba_stencil, x3, y3):
    """
    background   : BGR image (second image, e.g., pool table)
    rgba_stencil : RGBA image (ball stencil)
    (x3, y3)     : center position where stencil will be placed

    returns: composited BGR image
    """

    bg = background.copy()

    h_s, w_s, _ = rgba_stencil.shape
    h_bg, w_bg, _ = bg.shape

    # Top-left corner so that (x3, y3) is the center
    x1 = int(x3 - w_s / 2)
    y1 = int(y3 - h_s / 2)

    # Compute valid overlay region
    x_start = max(x1, 0)
    y_start = max(y1, 0)
    x_end   = min(x1 + w_s, w_bg)
    y_end   = min(y1 + h_s, h_bg)

    # Corresponding stencil region
    sx1 = x_start - x1
    sy1 = y_start - y1
    sx2 = sx1 + (x_end - x_start)
    sy2 = sy1 + (y_end - y_start)

    # Split stencil channels
    stencil_rgb = rgba_stencil[sy1:sy2, sx1:sx2, :3]
    alpha = rgba_stencil[sy1:sy2, sx1:sx2, 3] / 255.0
    alpha = alpha[..., np.newaxis]

    # Alpha blending
    bg[y_start:y_end, x_start:x_end] = (
        alpha * stencil_rgb +
        (1 - alpha) * bg[y_start:y_end, x_start:x_end]
    ).astype(np.uint8)

    return bg
````
