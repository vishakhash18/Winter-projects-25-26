```python
#now we will create a manual scaler
import numpy as np
def manual_standardscaler(X):
    mean = X.mean(axis=0)          # CRITICAL: axis=0
    std = X.std(axis=0)             # CRITICAL: axis=0
    std[std == 0] = 1.0            # Prevent division by zero
    X_scaled = (X - mean) / std
    return X_scaled, mean, std

X_scaled, mean, std = manual_standardscaler(X)
print("\nManual scaler result shape:", X_scaled.shape)
print("\nManual scaler result:\n", X_scaled)
  #we get a scaled value for the features which is represented by X
```

```python
import numpy as np
from sklearn.cluster import KMeans

def kmeans_matrix(M, K, random_state=42):
    """
    Args:
        M : 2D numpy array (H x W)
        K : number of clusters

    Returns:
        assignment_table : same shape as M, cluster labels
        cookbook : dict -> cluster_id : list of (i, j) positions
        centroids : numpy array of centroid values
    """
    H, W = M.shape

    # 1. Convert matrix to feature list
    # Each cell is one data point with 1 feature
    X = M.reshape(-1, 1)

    # 2. Run KMeans
    kmeans = KMeans(n_clusters=K, random_state=random_state, n_init=10)
    labels = kmeans.fit_predict(X)

    # 3. Assignment table (same shape as M)
    assignment_table = labels.reshape(H, W)

    # 4. Cookbook
    cookbook = {k: [] for k in range(K)}
    for i in range(H):
        for j in range(W):
            cluster_id = assignment_table[i, j]
            cookbook[cluster_id].append((i, j))

    # 5. Centroids
    centroids = kmeans.cluster_centers_

    return assignment_table, cookbook, centroids
```

```python
M = np.array([
    [1, 2, 3],
    [10, 11, 12],
    [2, 3, 4]
])

assignment_table, cookbook, centroids = kmeans_matrix(M, K=2)

print("Assignment Table:\n", assignment_table)
print("Cookbook:\n", cookbook)
print("Centroids:\n", centroids)
```
