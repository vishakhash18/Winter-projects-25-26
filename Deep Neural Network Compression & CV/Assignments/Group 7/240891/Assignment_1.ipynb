````python
rows , cols = fshift.shape
crow, ccol =  rows // 2, cols // 2
````

````python
def create_mask_lpf(fshift , radius):
  mask_lpf = np.zeros(fshift.shape, np.uint8)
  mask_lpf = cv2.circle(mask_lpf,(ccol,crow),radius,1,-1)

  return apply_mask(mask_lpf) , mask_lpf
````

````python
def apply_mask(mask): # Apply LPF Mask
  mask_lpf_norm = mask.astype(np.float32) / 255.0     # converting to float for further multiplication
  f_lpf = fshift*(mask_lpf_norm)                      # Applying Mask
  ishift_lpf = np.fft.ifftshift(f_lpf)                # Applting Ishift
  img_back_lpf = np.fft.ifft2(ishift_lpf)             # Applying FFT
  img_back_lpf = np.abs(img_back_lpf)

  return show_img(mask,np.log(abs(f_lpf)+1),img_back_lpf,"Mask","LPF","Inverse FFT") , img_back_lpf
````

````python
def show_img(img1,img2,img3,title1,title2,title3): # Combined Show Image function for better comparision
  plt.subplot(1,3,1)
  plt.imshow(img1,cmap="grey")
  plt.title(title1)
  plt.axis("off")

  plt.subplot(1,3,2)
  plt.imshow(img2,cmap="grey")
  plt.title(title2)
  plt.axis("off")

  plt.subplot(1,3,3)
  plt.imshow(img3,cmap="grey")
  plt.title(title3)
  plt.axis("off")

  plt.show()
````

````python
def apply_mask_hpf(mask_lpf): # Create & Apply HPF Mask
  mask_hpf=1-mask_lpf
  mask_hpf_norm = mask_hpf.astype(np.float32) / 255.0 # converting to float for further multiplication
  f_hpf = fshift*(mask_hpf)                      # Applying Mask
  ishift_hpf = np.fft.ifftshift(f_hpf)
  img_back_hpf = np.fft.ifft2(ishift_hpf)
  img_back_hpf = np.abs(img_back_hpf)

  return show_img(mask_hpf,np.log(np.abs(f_hpf)+1),img_back_hpf,"Mask","HPF","Inverse FFT")
````

In the Fourier domain, the central region contains low frequencies, which correspond to smooth variations and overall structure in the image. The outer regions contain high frequencies, which capture sharp edges, fine details, textures, and noise.

A circular mask preserves only the frequencies that fall within its radius.

Because the radius determines how many high-frequency components are removed, adjusting it directly controls the level of blur in the resulting image.

````python
#C3 Creating a square filter
x1 = ccol - 25
y1 = crow + 25
x2 = ccol + 25
y2 = crow - 25
mask_lpf_square = np.zeros(fshift.shape,np.uint8)
cv2.rectangle(mask_lpf_square, (x1, y1), (x2, y2), 255, -1)
````

````python
mask_lpf_norm = mask_lpf_square.astype(np.float32) / 255.0 # converting to float for further multiplication
f_lpf = fshift*(mask_lpf_norm)                      # Applying Mask
ishift_lpf = np.fft.ifftshift(f_lpf)                # Applting Ishift
img_back_lpf = np.fft.ifft2(ishift_lpf)             # Applying FFT
img_back_lpf = np.abs(img_back_lpf)

plt.subplot(1,2,1)
plt.imshow(np.log(np.abs(f_lpf)+1), cmap = "grey")  # Reconstructed Image
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_back_lpf, cmap = "grey")  # Reconstructed Image
plt.axis("off")
plt.show()
````

````python
#D1

#Reconstruct
phase_only = fshift / np.abs(fshift)
phase_ishift = np.fft.ifftshift(phase_only)
img_phase = np.fft.ifft2(phase_ishift)
img_phase = np.abs(img_phase)
#Display
plt.imshow(img_phase , cmap="grey")
plt.title("Phase_only")
plt.axis("off")
plt.show()
````

In this approach, each frequency component is normalized by its magnitude, meaning the magnitude is set to 1. This preserves the original phase information from the FFT but completely discards all magnitude information.

On the other hand, using np.angle() returns only the phase values, not the full complex numbers. To reconstruct the frequency components, those phase values must be converted back into complex exponentials.

````python
# Using np.angle(fshift)
phase_angle= np.angle(fshift)
phase_angle=np.exp(1j*phase_angle)

phase_angle_ishift = np.fft.ifftshift(phase_angle)
img_phase = np.fft.ifft2(phase_angle_ishift)
img_phase = np.abs(img_phase)

# Display image using np.angle()
plt.imshow(img_phase, cmap='gray')
plt.axis("off")
plt.show()
````

The output from both the approches is similar.

````python
#D2

#Reconstruct
mag_only = np.abs(fshift)

mag_ishift = np.fft.ifftshift(mag_only)
img_mag = np.fft.ifft2(mag_ishift)
img_mag = np.abs(img_mag)

#Display
plt.imshow(img_mag, cmap="grey")
plt.title("mag_only")
plt.axis("off")
plt.show()
````

This type of magnitude-only reconstruction results in almost the entire spectrum appearing black. The bright white blob visible in the shifted magnitude spectrum disappears after the inverse FFT because it shifts back to the corner, where it becomes just a single bright pixel rather than a central region.

Generally, Phase carries all the geometric informations.
