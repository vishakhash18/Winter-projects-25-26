To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# sobel kernel

sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

sobel_y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
], dtype=np.float32)
````

````python
def convolve(img, kernel):
    H, W = img.shape
    kH, kW = kernel.shape
    pad = kH // 2

    img_padded = np.pad(img, pad, mode='reflect')
    out = np.zeros_like(img, dtype=np.float32)

    for i in range(H):
        for j in range(W):
            region = img_padded[i:i+kH, j:j+kW]
            out[i,j] = np.sum(region * kernel)

    return out
````

````python
def analyze_image(img, threshold=None,title = " " ):

    # Sobel gradients
    Ix = convolve(img, sobel_x)
    Iy = convolve(img, sobel_y)

    # Thresholding
    if threshold is not None:
        Ix = np.where(np.abs(Ix) > threshold, Ix, 0)
        Iy = np.where(np.abs(Iy) > threshold, Iy, 0)

    # Plot gradients
    plt.figure(figsize=(15,4))

    plt.subplot(1,3,1)
    plt.imshow(Ix, cmap='gray')
    plt.title(title +" Ix")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(Iy, cmap='gray')
    plt.title(title + " Iy")
    plt.axis("off")

    # Scatter plot
    plt.subplot(1,3,3)
    plt.scatter(Ix.flatten(), Iy.flatten(), s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(title + " Iy vs Ix Scatter")

    plt.show()

    return Ix, Iy
````

````python
# Flat image
Ix_flat, Iy_flat = analyze_image(flat,20,"Flat")

# Edge image
Ix_edge, Iy_edge = analyze_image(edge,20,"Edge")

# Corner image
Ix_corner, Iy_corner = analyze_image(corner,20,"Corner")
````

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def compute_Ix_Iy(gray):
    gray = gray.astype(np.float32)

    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]], dtype=np.float32)

    sobel_y = np.array([[-1, -2, -1],
                        [ 0,  0,  0],
                        [ 1,  2,  1]], dtype=np.float32)

    Ix = cv2.filter2D(gray, -1, sobel_x)
    Iy = cv2.filter2D(gray, -1, sobel_y)

    return Ix, Iy
````

````python
def harris_R_map(gray, window_size=3, k=0.04):
    Ix, Iy = compute_Ix_Iy(gray)

    rows, cols = gray.shape
    offset = window_size // 2

    R_map = np.zeros((rows, cols), dtype=np.float32)

    for i in range(offset, rows - offset):
        for j in range(offset, cols - offset):

            Ix_win = Ix[i-offset:i+offset+1, j-offset:j+offset+1]
            Iy_win = Iy[i-offset:i+offset+1, j-offset:j+offset+1]

            # Second moments
            a = np.sum(Ix_win ** 2)
            b = np.sum(Ix_win * Iy_win)
            c = np.sum(Iy_win ** 2)

            # Harris response
            det = a * c - b * b
            trace = a + c
            R_map[i, j] = det - k * (trace ** 2)

    return R_map
````

````python
def show_R_map(gray, title="Harris R Map"):
    R = harris_R_map(gray)

    plt.figure(figsize=(6,6))
    plt.imshow(R, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()

    return R
````

````python
# display result
show_R_map(flat,   "Flat Image R Map")
show_R_map(edge,   "Edge Image R Map")
show_R_map(corner, "Corner Image R Map")
````

````python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('images (4).jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_gray,cmap='gray')
plt.title('Grayscale Image')
plt.axis("off")
plt.show()
````

````python
def draw_lines_m_c(img_gray, T):
    h, w = img_gray.shape

    # Canny edge detection
    edges = cv2.Canny(img_gray, 100, 200)

    # Edge coordinates
    ys, xs = np.where(edges > 0)

    # m and c ranges (kept reasonable)
    m_values = np.linspace(-3, 3, 400)
    c_values = np.linspace(-h, h, 400)

    Nm = len(m_values)
    Nc = len(c_values)

    # Accumulator
    acc = np.zeros((Nm, Nc), dtype=np.float32)

    dc = c_values[1] - c_values[0]

    # Voting
    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_values):
            c = y - m * x
            ci = int((c - c_values[0]) / dc)

            if 0 <= ci < Nc:
                acc[mi, ci] += 1

    # Normalize accumulator
    if acc.max() > 0:
        acc /= acc.max()

    # Threshold
    inds = np.where(acc >= T)
    detected_m = m_values[inds[0]]
    detected_c = c_values[inds[1]]

    # Draw lines
    for m, c in zip(detected_m, detected_c):
        x1, x2 = 0, w
        y1 = int(m * x1 + c)
        y2 = int(m * x2 + c)

        if 0 <= y1 < h and 0 <= y2 < h:
            cv2.line(img, (x1, y1), (x2, y2), (255, 0, 0), 1)

    # Display
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.title("Edge Map")
    plt.imshow(edges, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Accumulator (mc)")
    plt.imshow(acc, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Detected Lines (mc)")
    plt.imshow(img)
    plt.axis("off")

    plt.show()
````

````python
def draw_lines_r_theta(img_gray, T):
    h, w = img_gray.shape

    # 1. Canny edge detection
    edges = cv2.Canny(img_gray, 100, 200)

    # 2. Edge coordinates
    ys, xs = np.where(edges > 0)

    # 3. Rho and theta ranges
    D = int(np.sqrt(h*h + w*w))
    rho_values = np.linspace(-D, D, 400)
    theta_values = np.linspace(0, np.pi, 180)

    Nr = len(rho_values)
    Nt = len(theta_values)

    # 4. Accumulator
    acc = np.zeros((Nr, Nt), dtype=np.float32)

    dr = rho_values[1] - rho_values[0]

    # 5. Voting
    for x, y in zip(xs, ys):
        for ti, theta in enumerate(theta_values):
            rho = x * np.cos(theta) + y * np.sin(theta)
            ri = int((rho - rho_values[0]) / dr)

            if 0 <= ri < Nr:
                acc[ri, ti] += 1

    # 6. Normalize accumulator
    if acc.max() > 0:
        acc /= acc.max()

    # 7. Threshold
    inds = np.where(acc >= T)
    detected_rho = rho_values[inds[0]]
    detected_theta = theta_values[inds[1]]

    # 8. Draw detected lines
    for rho, theta in zip(detected_rho, detected_theta):
        a = np.cos(theta)
        b = np.sin(theta)

        x0 = a * rho
        y0 = b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(img, (x1, y1), (x2, y2), (255, 0, 0), 1)

    # 9. Display results
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.title("Edge Map")
    plt.imshow(edges, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Accumulator ()")
    plt.imshow(acc, cmap='gray', aspect='auto')
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Detected Lines ()")
    plt.imshow(img)
    plt.axis("off")

    plt.show()
````
