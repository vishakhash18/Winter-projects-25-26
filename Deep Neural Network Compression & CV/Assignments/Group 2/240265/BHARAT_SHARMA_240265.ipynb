````python
import numpy as np
import matplotlib.pyplot as plt

# Get image dimensions from the grayscale image (from Part A)
rows, cols = grayscale_image_np.shape
crow, ccol = rows // 2, cols // 2

# Create a meshgrid for frequency domain coordinates
x = np.arange(cols)
y = np.arange(rows)
X, Y = np.meshgrid(x, y)

# Calculate the distance from the center for each point
distance = np.sqrt((X - ccol)**2 + (Y - crow)**2)

### C1. Create 3 LPF masks
radii = [10, 30, 80]

for radius in radii:
    # Create a circular Low-Pass Filter (LPF) mask
    # The mask is 1 inside the given radius and 0 outside
    lpf_mask = distance < radius

    # Apply the LPF mask to the shifted FFT
    fshift_lpf = fshift * lpf_mask

    # Reconstruct the image using inverse FFT shift and inverse FFT
    f_ishift_lpf = np.fft.ifftshift(fshift_lpf)
    img_back_lpf = np.fft.ifft2(f_ishift_lpf)
    img_back_lpf = np.abs(img_back_lpf) # Take the absolute value for display

    # Normalize the image for display
    img_back_lpf = (img_back_lpf - np.min(img_back_lpf)) / (np.max(img_back_lpf) - np.min(img_back_lpf))

    # Display the reconstructed image
    plt.imshow(img_back_lpf, cmap='gray')
    plt.title(f'LPF Reconstructed Image (Radius: {radius})')
    plt.axis('off')
    plt.show()

# Explanation for blur levels:
# As the radius decreases, more high-frequency components (which correspond to fine details and edges)
# are blocked by the LPF mask. This results in a stronger blurring effect. Conversely, larger radii
# allow more high-frequency components to pass, leading to less blurring and a sharper image.

### C2. Create HPF mask
# For HPF, we choose a radius. Let's use 30 for demonstration, but it can be adjusted.
hpf_radius = 30

# Create a circular High-Pass Filter (HPF) mask
# The mask is 0 inside the given radius and 1 outside
hpf_mask = distance > hpf_radius

# Apply the HPF mask to the shifted FFT
fshift_hpf = fshift * hpf_mask

# Reconstruct the image
f_ishift_hpf = np.fft.ifftshift(fshift_hpf)
img_back_hpf = np.fft.ifft2(f_ishift_hpf)
img_back_hpf = np.abs(img_back_hpf)

# Normalize the image for display
img_back_hpf = (img_back_hpf - np.min(img_back_hpf)) / (np.max(img_back_hpf) - np.min(img_back_hpf))

# Display the reconstructed image
plt.imshow(img_back_hpf, cmap='gray')
plt.title(f'HPF Reconstructed Image (Radius: {hpf_radius})')
plt.axis('off')
plt.show()

### C3. Create a square LPF mask
side = 50

# Create a square mask with 1s at the center and 0s elsewhere
square_lpf_mask = np.zeros((rows, cols), dtype=bool)
# Calculate the start and end indices for the square region
start_row = crow - side // 2
end_row = crow + side // 2
start_col = ccol - side // 2
end_col = ccol + side // 2

# Ensure indices are within bounds (important for small images or large sides)
start_row = max(0, start_row)
end_row = min(rows, end_row)
start_col = max(0, start_col)
end_col = min(cols, end_col)

square_lpf_mask[start_row:end_row, start_col:end_col] = 1

# Apply the square LPF mask to the shifted FFT
fshift_square_lpf = fshift * square_lpf_mask

# Reconstruct the image
f_ishift_square_lpf = np.fft.ifftshift(fshift_square_lpf)
img_back_square_lpf = np.fft.ifft2(f_ishift_square_lpf)
img_back_square_lpf = np.abs(img_back_square_lpf)

# Normalize the image for display
img_back_square_lpf = (img_back_square_lpf - np.min(img_back_square_lpf)) / (np.max(img_back_square_lpf) - np.min(img_back_square_lpf))

# Display the reconstructed image
plt.imshow(img_back_square_lpf, cmap='gray')
plt.title(f'Square LPF Reconstructed Image (Side: {side})')
plt.axis('off')
plt.show()
````

````python
#D1. Phase-only reconstruction

# Calculate phase_only component
# This gives a complex array where the magnitude of each element is 1, and the phase is the original phase of fshift
phase_only = fshift / np.abs(fshift)

# Reconstruct the image from the phase-only component
f_ishift_phase_only = np.fft.ifftshift(phase_only)
img_back_phase_only = np.fft.ifft2(f_ishift_phase_only)
img_back_phase_only = np.abs(img_back_phase_only) # Take the absolute value for display

# Normalize and display
img_back_phase_only = (img_back_phase_only - np.min(img_back_phase_only)) / (np.max(img_back_phase_only) - np.min(img_back_phase_only))
plt.imshow(img_back_phase_only, cmap='gray')
plt.title('Phase-Only Reconstruction')
plt.axis('off')
plt.show()

# Comment 1: Difference between `fshift / np.abs(fshift)` and `np.angle(fshift)`
# `fshift / np.abs(fshift)`: This operation normalizes the magnitude of each complex number in `fshift` to 1, while preserving its original phase. The result is a complex array that retains all the phase information but discards the magnitude information.
# `np.angle(fshift)`: This function directly extracts only the phase (angle) component of each complex number in `fshift`, returning a real-valued array of angles in radians. It does not produce a complex frequency domain representation that can be directly used with `ifft2` to reconstruct an image in a meaningful way (as shown below).

# Try reconstruction from `np.angle(fshift)`
# When we reconstruct from `np.angle(fshift)` directly, we are essentially treating the phase values as if they were complex numbers with magnitude 1 and that angle. However, `np.angle` returns real numbers, not complex numbers suitable for direct inverse FFT. To attempt reconstruction, we can construct a complex array with magnitude 1 and phase from `np.angle(fshift)`.
phase_angles = np.angle(fshift)
reconstructed_from_angle = np.exp(1j * phase_angles)

f_ishift_from_angle = np.fft.ifftshift(reconstructed_from_angle)
img_back_from_angle = np.fft.ifft2(f_ishift_from_angle)
img_back_from_angle = np.abs(img_back_from_angle)

# Normalize and display
img_back_from_angle = (img_back_from_angle - np.min(img_back_from_angle)) / (np.max(img_back_from_angle) - np.min(img_back_from_angle))
plt.imshow(img_back_from_angle, cmap='gray')
plt.title('Reconstruction from np.exp(1j * np.angle(fshift))')
plt.axis('off')
plt.show()

# Comment 2: Explanation of output from `np.angle(fshift)` reconstruction
'''The output from reconstructing using `np.exp(1j * np.angle(fshift))` should be visually identical to the `phase_only` reconstruction.
This is because `fshift / np.abs(fshift)` is mathematically equivalent to `np.exp(1j * np.angle(fshift))`.
Both methods essentially create a complex spectrum where only the phase information is retained (with magnitudes normalized to 1), leading to the same reconstructed image that captures the structural information of the original image.'''
````

````python
#D2. Magnitude-only reconstruction

# Calculate magnitude_only component. Assuming zero phase for reconstruction, as is typical when isolating magnitude effects.
# We construct a complex array with the original magnitudes and zero phase.
mag_only_complex = np.abs(fshift) * np.exp(1j * np.zeros_like(fshift)) # Equivalent to just np.abs(fshift) if interpreted as real part and zero imaginary part, but explicitly makes it complex for ifft.

# Reconstruct the image from the magnitude-only component
f_ishift_mag_only = np.fft.ifftshift(mag_only_complex)
img_back_mag_only = np.fft.ifft2(f_ishift_mag_only)
img_back_mag_only = np.abs(img_back_mag_only) # Take the absolute value for display

# Normalize and display
img_back_mag_only = (img_back_mag_only - np.min(img_back_mag_only)) / (np.max(img_back_mag_only) - np.min(img_back_mag_only))
plt.imshow(img_back_mag_only, cmap='gray')
plt.title('Magnitude-Only Reconstruction')
plt.axis('off')
plt.show()

# Comment 3: Role of logarithm in Magnitude Spectrum and why this looks different.
# The logarithm is used in the `magnitude_spectrum = 20 * np.log(np.abs(fshift) + 1)` calculation primarily for visualization. The magnitudes of FFT coefficients can vary over a very wide range, often spanning several orders of magnitude. A direct plot of `np.abs(fshift)` would show a bright central spot (low frequencies) and very faint or invisible high-frequency components.
# By applying a logarithm (like `np.log`), this vast dynamic range is compressed, making it possible to visualize both the very large low-frequency components and the much smaller high-frequency components more clearly on a single display. The `+1` is added to prevent `log(0)` errors and to map zero values to a discernible level.
# The `Magnitude-Only Reconstruction` shown here, on the other hand, is the result of applying the inverse FFT *directly* to `np.abs(fshift)` (with zero phase). It is not a visualization of the frequency domain itself, but rather the spatial domain image reconstructed using only magnitude information. It generally produces a blurred, texture-like image without clear structural details, as the crucial phase information for spatial coherence is missing.
````
