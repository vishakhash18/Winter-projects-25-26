```python
s = "Python Assignment"

s_upper = s.upper()
print(s_upper)

s_reversed = ""
for i in range(len(s) - 1, -1, -1):
    s_reversed += s[i]
print(s_reversed)

vowels = "aeiouAEIOU"
vowel_count = 0
for(char) in s:
    if char in vowels:
        vowel_count += 1
print(vowel_count)
```

### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```
