### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

test_numbers = [7, 15, 23]
for num in test_numbers:
    print(f"Is {num} prime? {is_prime(num)}")
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]

# 1. Sort without using sort() - Implementing Bubble Sort
sorted_nums = numbers.copy()
n = len(sorted_nums)
for i in range(n):
    for j in range(0, n-i-1):
        if sorted_nums[j] > sorted_nums[j+1]:
            # Swap elements
            sorted_nums[j], sorted_nums[j+1] = sorted_nums[j+1], sorted_nums[j]

print("Sorted list:", sorted_nums)

# 2. Sum of all elements
total_sum = 0
for x in numbers:
    total_sum += x
print("Sum of elements:", total_sum)

# 3. Remove all odd numbers
# We create a new list keeping only even numbers
evens_only = [x for x in numbers if x % 2 == 0]
print("Remaining list (evens only):", evens_only)
```

### Q13. Line Plot

**Plot a simple line graph for:**

x = [1, 2, 3, 4, 5]

y = [2, 4, 6, 8, 10]

Add title and labels.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', color='blue') # Added marker for clarity
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```
