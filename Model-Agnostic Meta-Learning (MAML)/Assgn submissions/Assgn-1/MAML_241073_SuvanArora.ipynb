### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def check_prime(n) :
  if n<=1 :
    return False

  for i in range(2,int(n**0.5)+1) :
    if n%i == 0 :
      return False
  return True

print("Is 7 prime : ",check_prime(7))
print("Is 15 prime : ",check_prime(15))
print("Is 23 prime : ",check_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5,2,9,1,7,3]
n = len(numbers)

for i in range(n):
  for j in range(i+1,n):
    if numbers[i] > numbers[j]:
      temp = numbers[j]
      numbers[j] = numbers[i]
      numbers[i] = temp


print("Sorted list is : ",numbers)

sum = 0
for i in range(n):
  sum += numbers[i]

print("Sum of all elements is : ",sum)

even_only = []
for i in range(n):
  if numbers[i]%2 == 0 :
    even_only.append(numbers[i])

print("List with only even elements is : ",even_only)
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.random.normal(loc = 50, scale = 10, size = 100)

plt.hist(random_numbers, bins=10, color='blue', edgecolor='black')
plt.title("Histogram of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.plot()
print("Here a normal frequency distribution is considered with values ranging from 1-100")
print("The required histogram is plotted below : \n")
plt.show()
```
