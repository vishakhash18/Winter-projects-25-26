### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
    # Prime numbers are greater than 1
    if n < 2:
        return False
    # Check for factors from 2 up to the square root of n
    # If n is divisible by any number in this range, it's not prime
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Test cases for the is_prime function
print(f"check for 7: {is_prime(7)}") # Expected: True (7 is a prime number)
print(f"check for 15: {is_prime(15)}") # Expected: False (15 is divisible by 3 and 5)
print(f"check for 23: {is_prime(23)}") # Expected: True (23 is a prime number)
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]

n = len(numbers)
# Bubble sort algorithm to sort the list without using the built-in sort() method
for i in range(n - 1):
    for j in range(0, n - i - 1):
        # Compare adjacent elements and swap them if they are in the wrong order
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
print("Sorted list: ", numbers)

# Calculate the sum of all elements in the list
total_sum = 0
for num in numbers:
    total_sum += num
print("Sum:", total_sum)

# Create a new list containing only even numbers (removing odd numbers)
even_numbers = []
for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)
print("List after removing odd numbers:", even_numbers)
```

### Q9. NumPy Indexing

**Create an array of numbers 1–20.**
Extract:
- All even numbers
- Values from index 5 to 10
- Every 3rd element

```python
import numpy as np

# Create an array of numbers from 1 to 20
numbers_array = np.arange(1, 21)
print("Original array:", numbers_array)
print("\n")

# Extract all even numbers from the array using boolean indexing
even_numbers = numbers_array[numbers_array % 2 == 0]
print("Even numbers:", even_numbers)
print("\n")

# Extract values from index 5 to 10 (inclusive of 5, inclusive of 10) using slicing
values_5_to_10 = numbers_array[5:11]
print("Values from index 5 to 10:", values_5_to_10)
print("\n")

# Extract every 3rd element from the array using slicing with a step
every_3rd_element = numbers_array[::3]
print("Every 3rd element:", every_3rd_element)
```

### Q14. Bar Chart

**Plot a bar chart of 4 subjects vs marks:**
- Maths – 90
- Physics – 75
- Chemistry – 88
- CS – 95

```python
import matplotlib.pyplot as plt

subjects = ['Maths', 'Physics', 'Chemistry', 'CS']
marks = [90, 75, 88, 95]

plt.figure(figsize=(8, 6))
plt.bar(subjects, marks, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks by Subject')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--')
plt.show()
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random numbers
random_numbers = np.random.randn(100)

# Plot a histogram
plt.hist(random_numbers, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Random Numbers')
plt.grid(axis='y', linestyle='--')
plt.show()
```
