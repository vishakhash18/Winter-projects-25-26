### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
import math

def is_prime(n):
  if n == 1 or n == 0:
    return False
  for i in range(2, math.ceil((n ** 0.5) + 1)):
    if(n % i == 0):
        return False
  return True

print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]
n = len(numbers)
for i in range(n):
  for j in range(i):
    if numbers[i] < numbers[j]:
      numbers[i], numbers[j] = numbers[j], numbers[i]
sum = 0
for i in numbers:
  sum += i
numbers = [x for x in numbers if x % 2 == 0]
```
